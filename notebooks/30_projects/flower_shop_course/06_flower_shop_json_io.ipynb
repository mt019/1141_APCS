## 1. 寫出庫存到 JSON

```python
import json

inventory = {
    "玫瑰": {"cost": 15, "price": 30, "stock": 20},
    "百合": {"cost": 18, "price": 36, "stock": 12},
}

with open("inventory.json", "w", encoding="utf-8") as f:
    json.dump(inventory, f, ensure_ascii=False, indent=2)
```

## 2. 讀回 JSON 還原庫存

```python
import json

with open("inventory.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)

loaded
```

## 3. 訂單明細（巢狀 JSON）

```python
import json

orders = [
    {
        "id": "OD001",
        "items": [{"name": "玫瑰", "qty": 2}, {"name": "百合", "qty": 1}],
        "total": 120,
        "status": "paid",
    },
    {
        "id": "OD002",
        "items": [{"name": "康乃馨", "qty": 3}],
        "total": 60,
        "status": "created",
    },
]

with open("orders.json", "w", encoding="utf-8") as f:
    json.dump(orders, f, ensure_ascii=False, indent=2)
```

## 4. 查詢與更新

```python
import json

with open("orders.json", "r", encoding="utf-8") as f:
    orders = json.load(f)

# 找出還沒付款的訂單
pending = [o for o in orders if o["status"] != "paid"]

# 更新一筆狀態
orders[1]["status"] = "paid"

pending, orders[1]
```
