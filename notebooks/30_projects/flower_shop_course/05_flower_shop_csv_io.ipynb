## 1. 寫出庫存到 CSV（DictWriter）

```python
import csv

inventory = {
    "玫瑰": {"cost": 15, "price": 30, "stock": 20},
    "百合": {"cost": 18, "price": 36, "stock": 12},
}

with open("inventory.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "cost", "price", "stock"])
    writer.writeheader()
    for name, info in inventory.items():
        writer.writerow({"name": name, **info})
```

## 2. 讀回 CSV 還原庫存（DictReader）

```python
import csv

loaded = {}
with open("inventory.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        loaded[row["name"]] = {
            "cost": int(row["cost"]),
            "price": int(row["price"]),
            "stock": int(row["stock"]),
        }

loaded
```

## 3. 記錄訂單到 CSV（追加寫入）

```python
import csv

orders = [
    {"date": "2026-02-01", "order_id": "OD001", "total": 120},
    {"date": "2026-02-01", "order_id": "OD002", "total": 260},
    {"date": "2026-02-02", "order_id": "OD003", "total": 180},
]

with open("orders.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["date", "order_id", "total"])
    writer.writeheader()
    for o in orders:
        writer.writerow(o)
```

## 4. 簡易報表：指定日期營收

```python
import csv

def revenue_by_date(date):
    total = 0
    with open("orders.csv", "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row["date"] == date:
                total += int(row["total"])
    return total

revenue_by_date("2026-02-01")
```
