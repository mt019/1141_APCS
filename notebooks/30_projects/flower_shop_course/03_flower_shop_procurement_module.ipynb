## 1. 用 dataclass 定義花材與庫存

```python
from dataclasses import dataclass

@dataclass
class Item:
    name: str
    cost: int
    price: int

@dataclass
class Stock:
    item: Item
    qty: int

rose = Item("玫瑰", 15, 30)
rose_stock = Stock(rose, 20)
rose_stock
```

## 2. 庫存模組：進貨 / 補貨

```python
def add_new_item(inventory, item, qty):
    # 新品項入庫
    if item.name in inventory:
        return False
    inventory[item.name] = Stock(item, qty)
    return True


def restock(inventory, name, qty):
    # 老品項補貨
    if name in inventory:
        inventory[name].qty += qty
        return True
    return False

inventory = {}
add_new_item(inventory, Item("玫瑰", 15, 30), 20)
restock(inventory, "玫瑰", 5)

inventory["玫瑰"]
```

## 4. 結構化打印：庫存清單

```python
def print_inventory(inv):
    print("=== 今日庫存 ===")
    print("NAME	COST	PRICE	QTY")
    for name, stock in inv.items():
        print(f"{name}	{stock.item.cost}	{stock.item.price}	{stock.qty}")

print_inventory(inventory)
```
