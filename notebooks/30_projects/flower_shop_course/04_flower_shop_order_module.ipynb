## 1. 訂單資料結構（dataclass）

```python
from dataclasses import dataclass

@dataclass
class Order:
    order_id: str
    items: list  # list of (name, qty)
    total: int = 0
    status: str = "created"
    remark: str = ""
    pickup: str = "自取"
```

## 2. 訂單建立與狀態更新

```python
def create_order(order_id, items, remark="", pickup="自取"):
    return Order(order_id, items, remark=remark, pickup=pickup)


def update_status(order, status):
    order.status = status
    return order

order = create_order("OD004", [("玫瑰", 1)], "一般")
update_status(order, "paid")
order
```

## 3. 訂單金額計算

```python
# inventory 使用簡化 dict
inventory = {
    "玫瑰": {"price": 30},
    "百合": {"price": 36},
}


def calc_total(inv, items):
    total = 0
    for name, qty in items:
        total += inv[name]["price"] * qty
    return total

order.total = calc_total(inventory, order.items)
order
```

```python
# TODO
order2 = create_order("OD005", [("百合", 2)], "一般", "外送")
order2.total = calc_total(inventory, order2.items)
order2
```

## 4. 結構化打印：標題 + 對齊欄位

```python
def print_order(order):
    print(f"ORDER	{order.order_id}")
    print(f"STATUS	{order.status}")
    print(f"PICKUP	{order.pickup}")
    print(f"REMARK	{order.remark}")
    print("ITEM	QTY")
    for name, qty in order.items:
        print(f"{name}	{qty}")
    print(f"TOTAL	{order.total}")

print_order(order2)
```

## 5. 訂單列表與統計（含結構化輸出）

```python
orders = []
orders.append(order)
orders.append(order2)

revenue = sum(o.total for o in orders)
print("ID	STATUS	TOTAL")
for o in orders:
    print(f"{o.order_id}	{o.status}	{o.total}")
print(f"REVENUE	{revenue}")
```

### 小練習 5
寫一個 `count_status`，統計各狀態訂單數量。

```python
# TODO

def count_status(orders):
    result = {}
    for o in orders:
        result[o.status] = result.get(o.status, 0) + 1
    return result

count_status(orders)
```
