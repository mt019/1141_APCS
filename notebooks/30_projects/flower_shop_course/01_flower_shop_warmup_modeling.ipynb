## 2. 常用結構：快速查詢
把 list 改成 **dict of dict**（名字當 key），查找更快。

```python
# key = 花名，value = 花的資訊
inventory = {
    "玫瑰": {"cost": 15, "price": 30, "stock": 20},
    "百合": {"cost": 18, "price": 36, "stock": 12},
    "康乃馨": {"cost": 10, "price": 20, "stock": 25},
}

inventory["玫瑰"]["stock"]
```

## 3. tuple / set 的位置
- **tuple**：固定格式的操作輸入（不改內容）
- **set**：花種去重、快速判斷存在

```python
# tuple：一次操作指令 (name, qty)
order = ("玫瑰", 3)

# set：所有花種（去重）
flower_types = {"玫瑰", "百合", "康乃馨", "玫瑰"}

order, flower_types
```

## 4. 把流程封裝成函式

```python
def add_stock(inv, name, qty):
    if name in inv:
        inv[name]["stock"] += qty
        return True
    return False


def sell_flower(inv, name, qty):
    if name in inv and inv[name]["stock"] >= qty:
        inv[name]["stock"] -= qty
        revenue = inv[name]["price"] * qty
        profit = (inv[name]["price"] - inv[name]["cost"]) * qty
        return True, revenue, profit
    return False, 0, 0

print(add_stock(inventory, "百合", 3))
print(sell_flower(inventory, "玫瑰", 4))
```
