## 1. 格式化輸出三種方式
在 Python 中常見的輸出字串方式有三種：`str.format`、`f-string` 以及 `string.Template`。先閱讀下方範例，觀察括號的使用與語法差異。


```python
# 1. str.format 基礎
name = "Amy"
score = 92
subject = "Math"

print("Student: {} | Score: {}".format(name, score))
print("Student: {name} | Score: {score}".format(name=name, score=score))
print("{:<10} | {:>5}".format(subject, score))
```

`str.format` 的重點：
- 透過 `{}` 佔位符插入資料，支援索引或名稱。
- 可搭配格式說明，例如 `{:>5}` 代表右對齊、`{:.2f}` 代表小數點後兩位。
- 適合需要重複套用同一份模版、或維護舊程式碼時使用。


```python
# 2. f-string（Python 3.6+）
print(f"Student: {name} | Score: {score}")
print(f"{name:>10} | {subject:^6} | {score:05d}")
print(f"加權後分數：{(score * 1.05):.1f}")

# 可以直接放入運算或函式
from statistics import mean
scores = [88, 92, 79]
print(f"Average score -> {mean(scores):.2f}")
```

`f-string` 的亮點：
- 變數直接寫在字串中，語意清楚。
- 支援即時運算與函式呼叫，適合撰寫新程式或快速除錯。
- 保留與 `str.format` 相同的格式化語法。


### `string.Template`
`string.Template` 位於 `string` 模組，適合需要讓使用者自訂模版、或避免 `{}` 語法衝突的情境。它使用 `$變數名` 作為佔位符。


```python
from string import Template

report_tpl = Template("""Hello $name,
本週課程主題：$topic
你的目前分數是 $score 分。
Keep going!
""")

message = report_tpl.substitute(name="Amy", topic="Loops & Brackets", score=92)
print(message)

# safe_substitute 可以在缺少欄位時避免程式崩潰
partial = Template("User: $user | Bonus: $bonus")
print(partial.safe_substitute(user="Ben"))  # 缺少 bonus 也不會報錯
```

`string.Template` 的適用場合：
- 佔位符簡潔、易於觀察，適合作為外部設定檔或讓非程式人員編輯的模版。
- 若字串本身含有大量 `{}`（例如 JSON），可避免與 `str.format` 衝突。
- `safe_substitute` 能在缺少資料時提供預設處理，不會直接噴錯。


### 三種方式比較


| 功能點 | `str.format` | `f-string` | `string.Template` |
| --- | --- | --- | --- |
| 語法 | `{}` 佔位符，支援索引/名稱 | 字串前加 `f`，變數直寫 | `$name` 形式佔位符 |
| 版本需求 | Python 2.7+、3.x | Python 3.6+ | Python 2.4+ |
| 動態運算 | 需先備變數或呼叫函式 | 可直接寫運算或函式 | 僅支援替換值，不可直接運算 |
| 模版共享 | 需要保留格式字串 | 需包含於程式碼 | 可存在外部檔案交給使用者修改 |
| 適合情境 | 舊程式、批次輸出 | 新程式、除錯、內嵌運算 | 使用者可編輯的信件、設定模版 |

## 2. 常見資料結構與括號
在 Python 中，不同括號對應不同資料結構。以下示例總結 list、tuple、set、dict 的特性與常用操作。


| 結構 | 使用括號 | 是否可變 | 是否保序 | 是否允許重複 | 常見用途 |
| --- | --- | --- | --- | --- | --- |
| list | `[]` | ✅ | ✅ | ✅ | 需要增刪、排序或依序處理的資料 |
| tuple | `()` | ❌ | ✅ | ✅ | 固定設定、函式多值回傳、字典鍵 |
| set | `{}` | ✅ | ❌ | ❌ | 去除重複、集合運算、快速成員檢查 |
| dict | `{}`（鍵值對） | ✅ | 3.7+ 迭代有序 | key 不重複 | 建立映射、查表 |


```python
# list 範例
scores = [85, 73, 92, 73]
scores.append(88)
scores.extend([91, 100])
print("Scores:", scores)
print("前三名:", scores[:3])
print("去重後排序:", sorted(set(scores)))

# tuple 拆包
coords = (25.03, 121.56)
x, y = coords
print(f"座標 -> x={x}, y={y}")

# set 交集/聯集
monday = {"Amy", "Ben", "Clara", "David"}
wednesday = {"Ben", "Clara", "Ella"}
print("共同出席:", monday & wednesday)
print("只出一次:", monday ^ wednesday)

# dict 操作
data = {"name": "Amy", "score": 92, "tags": ["loop", "formatting"]}
data["score"] += 3
data.setdefault("email", "尚未提供")
for key, value in data.items():
    print(f"{key:>6}: {value}")
```

## 3. 推導式與生成器深度解析
推導式（Comprehension）提供了使用括號快速建立資料的方式，依照外層括號不同，可產生 list、set、dict 或 generator。


### 3.1 推導式回顧
下例展示四種常見推導式：


```python
nums = range(1, 11)

# list comprehension：保留偶數平方
list_comp = [n ** 2 for n in nums if n % 2 == 0]

# set comprehension：去除重複並保留結果
set_comp = {word.lower() for word in ["Hello", "World", "HELLO"]}

# dict comprehension：建立映射
names = ["Amy", "Ben", "Clara"]
heights = [160, 172, 155]
dict_comp = {name: height for name, height in zip(names, heights)}

print("list_comp:", list_comp)
print("set_comp:", set_comp)
print("dict_comp:", dict_comp)
```

推導式的重要觀念：
- 放在前方的是生成的資料；`for` 與 `if` 用於篩選與轉換。
- 可視為 `for` 迴圈 + `append` 的縮寫，在保留可讀性的前提下讓程式更精煉。
- 仍應注意不要把過多邏輯塞入同一行，避免降低可維護性。


### 3.2 生成器（Generator）
生成器是「惰性」的一種資料來源：不會一次建立完整序列，而是在需要時才產生下一個值。


```python
# generator expression 與 list comprehension 的差異
nums = range(1, 11)
list_comp = [n * n for n in nums if n % 2 == 0]
gen_expr = (n * n for n in nums if n % 2 == 0)

print("List comprehension ->", list_comp)
print("Generator expression ->", gen_expr)  # 只是物件

print("第一個值:", next(gen_expr))
print("剩餘總和:", sum(gen_expr))

try:
    next(gen_expr)
except StopIteration:
    print("生成器已耗盡，無法再次取值。重新建立才行。")
```

生成器的關鍵特性：
- 惰性評估（lazy evaluation）：僅在被取用時才計算下一個值。
- 記憶體效率高：不需要儲存整個序列，適合處理大量或無限資料流。
- 單次遍歷：一旦走訪完畢即耗盡，若要再次使用須重新建立。

生成器可透過 `yield` 關鍵字建立：


```python
def countdown(start):
    while start > 0:
        yield start
        start -= 1

for number in countdown(5):
    print("Tick", number)

# 也可以搭配 next() 手動取得
count = countdown(3)
print("Manual next ->", next(count))
print("Manual next ->", next(count))
```

### 3.3 推導式 vs 生成器：何時使用？
- 如果需要立即使用完整序列、或要多次遍歷資料，選擇推導式（例如 list comprehension）。
- 如果只需要逐步取值、或資料量巨大，選擇生成器（generator expression / yield 函式）。
- 兩者皆可搭配 `if` 條件；善用括號區分不同結構。
