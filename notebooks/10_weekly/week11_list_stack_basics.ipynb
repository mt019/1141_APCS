## 1. 取值與切片
- `lst[i]`、`lst[-1]`、`lst[a:b]`、`lst[a:b:c]`。
- 更新元素：`lst[i] = value`；切片賦值會改變長度。


```python
lst = [10, 20, 30, 40, 50]
print('lst[0]', lst[0])
print('lst[-1]', lst[-1])
print('lst[1:4]', lst[1:4])

lst[2] = 99
lst[1:3] = [7, 8, 9]  # 會改變長度
print('after updates:', lst)
```

## 2. Stack 常用操作
- push：`lst.append(x)`。
- pop：`lst.pop()`（預設彈出尾端）；`lst.pop()` 空列表會丟出錯誤。
- peek：`lst[-1]` 但先確定非空。


```python
stack = []
stack.append('A')
stack.append('B')
stack.append('C')
print('stack now:', stack)

last = stack.pop()
print('popped:', last, 'now:', stack)

if stack:
    print('peek:', stack[-1])
```

```python
def push(stack, item):
    """將 item 推入 stack，回傳同一個 stack 便於串接。"""
    stack.append(item)
    return stack

demo = []
push(demo, 'X')
push(demo, 'Y')
print('after push:', demo)
```

### push / pop 動態範例
- 餐盤疊：先進後出，觀察每一步的 stack。
- Undo 類操作：把動作推入 stack，`pop` 代表撤銷最後一步。


```python
def log(action, stack):
    print(f"{action:<12} -> {stack}")

plates = []
for plate in ['A', 'B', 'C']:
    plates.append(plate)
    log(f'push {plate}', plates)

pop_one = plates.pop()
log(f'pop {pop_one}', plates)

undo_stack = []
actions = ['type A', 'type B', 'delete']
for act in actions:
    undo_stack.append(act)
    log(f'push {act}', undo_stack)

last = undo_stack.pop()
log(f'undo {last}', undo_stack)
```

## 3. 巢狀迴圈遍歷（list of list）
- `for r, row in enumerate(grid): for c, val in enumerate(row): ...`
- 小心列長度不一致時，取值前先確認或使用 `len(row)`。


```python
grid = [
    [1, 2, 3],
    [4, 5],
    [6, 7, 8, 9],
]
coords = []
for r, row in enumerate(grid):
    for c, val in enumerate(row):
        coords.append((r, c, val))
coords[:6]
```

## 4. 常見陷阱
- 淺拷貝：`alias = lst` 會共用；要複製用 `list(lst)` 或切片。
- 重複引用：`bad = [[0]*3]*2` 會共用同一列。
- `pop` 空列表會 `IndexError`，先檢查長度。


```python
origin = [1, 2, 3]
alias = origin
copy1 = list(origin)

alias[0] = 99
print('origin after alias change:', origin)

bad = [[0]*3]*2
bad[0][1] = 7
print('bad shared rows:', bad)

try:
    [].pop()
except Exception as e:
    print('pop on empty:', type(e).__name__)
```
