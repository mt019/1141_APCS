## 語法鋪墊：tuple 與 list 重點
- 逐層拆解：`students = [(name, [score1, score2, score3]), ...]`，先找出 tuple、再看內層 list。
- 拆封技巧：`for name, scores in students:` → `name` 是字串、`scores` 是 list；`first, *rest = scores` 幫助切開。
- 切片預設：`scores[:2]` 等同 `scores[0:2]`，`scores[1:]` 等同 `scores[1:len(scores)]`。遇到冒號省略時，先補上預設值再思考。
- 搭配表格：先寫出 `name`、`scores`、`sum(scores)`、`len(scores)`，最後才寫 comprehension。


```python
# 範例：tuple 與 list 的基本操作
point = (25.0478, 121.5319)
lat, lng = point
print('lat =', lat, '| lng =', lng)

scores = [82, 91, 88]
scores.append(95)
front = scores[0:2]
tail = scores[-2:len(scores)]
print('scores =', scores)
print('front =', front, '| tail =', tail)

students = [('Amy', [82, 91, 88]), ('Ben', [70, 76, 80])]
# for name, records in students:
for student in students:
    name, records = student
    avg = round(sum(records) / len(records), 1)
    print(f'{name} -> 平均 {avg}')
```
