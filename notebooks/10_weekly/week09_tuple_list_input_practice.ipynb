## 1. strip / split 練習

很多 APCS 題目會說（例如題庫 `APCS-程式實作初級題本範例.pdf` 裡）：

> 每筆測試資料之第一行有一整數 n，第二行包含 n 個以空白分隔之整數 …

這種描述幾乎都可以用同一套流程處理：
1. 用 `strip()` 把末尾的換行符號去掉。
2. 再用 `split()` 依空白切成 **字串 list**。

````python
line = "  10 20 30 40  \n"  # 模擬從 input() 讀進來的一行
print(repr(line))  # 用 repr 看得出換行與空白

clean = line.strip()
print("strip 之後:", repr(clean))

parts = clean.split()
print("split 之後:", parts, "型態:", type(parts))
````

## 2. map(int, ...)：把字串 list 變成 int list

`split()` 之後得到的是 **字串**，但 APCS 題目常常要你做加總或比較大小，

````python
parts = ["10", "20", "30"]
nums = list(map(int, parts))
print("原本 parts:", parts, type(parts[0]))
print("轉成 nums:", nums, type(nums[0]))
````

## 最後小結

- `strip()`：先把多餘空白 / 換行去掉。
- `split()`：把一行字串切成一個個「小塊」，通常是 list[str]。
- `map(int, ...)`：把每個小塊轉成整數，變成 list[int]。
- tuple / list 結構：外層 list 裡面裝很多 `(name, scores)` / `(line1, line2)` 這樣的 tuple，裡面再裝 list[int]。
