## 七言對聯規則 Checklist
| 規則 | 說明 | 檢查方式 |
| --- | --- | --- |
| A：二四不同、二六同 | 第一句第2≠第4，且第2=第6；第二句同理 | `line[i]` 比較 |
| B：仄起平收 | 第一句第7要 1 (仄)，第二句第7要 0 (平) | 直接比較索引6 |
| C：相對性 | 第一句第2/4/6 與第二句對應位置相反 | `value1 != value2` |

### 範例：將規則轉成函式
```python
def check_rules(line1, line2):
    failed = []
    if not ((line1[1] != line1[3]) and (line1[1] == line1[5]) and (line2[1] != line2[3]) and (line2[1] == line2[5])):
        failed.append('A')
    if not (line1[6] == 1 and line2[6] == 0):
        failed.append('B')
    for idx in (1, 3, 5):
        if line1[idx] == line2[idx]:
            failed.append('C')
            break
    return failed or ['None']
```

### 小練習 2：實作 `check_rules`
- 依照上方規則完成函式，回傳排序好的規則字母或 `['None']`。

````python
def check_rules(line1, line2):
    failed = []
    if not ((line1[1] != line1[3]) and (line1[1] == line1[5]) and (line2[1] != line2[3]) and (line2[1] == line2[5])):
        failed.append('A')
    if not (line1[6] == 1 and line2[6] == 0):
        failed.append('B')
    for idx in (1, 3, 5):
        if line1[idx] == line2[idx]:
            failed.append('C')
            break
    return failed or ['None']

line1 = [1, 0, 0, 0, 0, 1, 1]
line2 = [0, 1, 0, 0, 1, 1, 0]
print('預期：[''A'', ''C''] =>', check_rules(line1, line2))
````
