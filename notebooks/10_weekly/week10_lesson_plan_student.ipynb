## 2.1 題庫：最大公因數 GCD (Greatest Common Divisor) 與輾轉相除法

題庫裡常見敘述：

> 給定兩正整數 a 與 b，請輸出它們的最大公因數。

先用「列出所有因數」直觀寫法，讓你感受兩種寫法的效率差異，再引入輾轉相除法：

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

````python
def gcd_slow(a: int, b: int) -> int:
    """列舉因數版本（給你當對照用）。"""
    ans = 1
    for d in range(1, min(a, b) + 1):
        if a % d == 0 and b % d == 0:
            ans = d
    return ans

def gcd_euclid(a: int, b: int) -> int:
    """輾轉相除法版本。"""
    while b != 0:
        a, b = b, a % b
    return a

print(gcd_slow(12, 18), gcd_euclid(12, 18))
````

### gcd_euclid 計算範例（18, 12）

| step | a | b | a % b | 說明 |
|------|---|---|-------|------|
| 0    | 18| 12| 6     | 一開始的 a, b |
| 1    | 12| 6 | 0     | 交換後：a=12, b=18 % 12 = 6 |
| 2    | 6 | 0 | —     | 交換後：a=6,  b=12 % 6 = 0 → 迴圈結束 |

最後得到的 a = 6，就是 GCD (Greatest Common Divisor)。

### 語法鋪墊：多重指定與 for 迴圈

在 Fibonacci 程式裡會看到這一行：`a, b = b, a + b`。

**IMPORTANT:** 這不是先改 a 再改 b，而是**同時**把右邊算好的兩個值，指定到左邊的兩個變數。
- 先算出右邊 `(b, a + b)`
- 再一次存回 `(a, b)`

也會用到 `for _ in range(2, n + 1):`：
- 這裡的 `_` 代表「我不在意這個變數的名字，只是想重複 n-1 次」。

### fib_iter 計算範例（n = 5）

一開始：`a = 0` (F0), `b = 1` (F1)。每跑一次迴圈就更新成下一對 Fibonacci。

| i (loop) | a (舊 F) | b (新 F) | 備註 |
|----------|---------|---------|------|
| 初始     | 0       | 1       | F0 = 0, F1 = 1 |
| 2        | 1       | 1       | F2 = 1 |
| 3        | 1       | 2       | F3 = 2 |
| 4        | 2       | 3       | F4 = 3 |
| 5        | 3       | 5       | F5 = 5 （函式回傳值） |


````python
def fib_iter(n: int) -> int:
    """回傳第 n 項 Fibonacci（以 F0 = 0, F1 = 1 起算）。"""
    if n == 0:
        return 0
    if n == 1:
        return 1
    a, b = 0, 1  # F0, F1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

for i in range(8):
    print(f"F{i}", fib_iter(i))
````
