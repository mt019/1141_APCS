## 0-1. 題目原文整理

### 問題描述

考慮一個二維正整數陣列 `A`，有 `n` 列 row 及 `m` 行 column。陣列 `A` 中第 `i` 列第 `j` 行的元素記為 `A[i][j]`。

兩個元素 `A[i][j]` 與 `A[s][t]` 的距離定為：

```python
abs(i - s) + abs(j - t)
```

對於一個元素 `A[i][j] = x`，若與它距離不超過 `x` 的元素總和，包含 `A[i][j]` 本身，除以 10 的餘數與 `x` 除以 10 的餘數相同，則稱 `(i, j)` 為特殊位置。

也就是說，若：

```python
total % 10 == x % 10
```

則 `(i, j)` 是特殊位置。

注意：計算總和時，超出陣列範圍的元素不計算。

### 輸入格式

第一行有兩個正整數 `n` 與 `m`，代表陣列列數與行數：

```text
1 <= n, m <= 50
```

接下來有 `n` 行，每行有 `m` 個整數。陣列內的數字均為小於 10 的正整數。

### 輸出格式

第一行輸出特殊位置總數 `k`。接下來 `k` 行，每行輸出一個特殊位置的列註標與行註標。

輸出順序：列註標小的先輸出；若列註標相同，行註標小的先輸出。這就是 row-major 順序。

## 10. 方框枚舉 + 距離篩選

這題可以先把候選範圍縮到方框：

```python
for s in range(r - x, r + x + 1):
    for t in range(c - x, c + x + 1):
        ...
```

接著用 `if` 檢查兩件事：

1. `(s, t)` 沒有出界。
2. `abs(r - s) + abs(c - t) <= x`。

這種寫法的重點是：先抓出可能的位置，再用條件排除出界或距離太遠的位置。

````python
def area_sum_box(grid, r, c):
    n = len(grid)
    m = len(grid[0])
    x = grid[r][c]
    total = 0

    for s in range(r - x, r + x + 1):
        for t in range(c - x, c + x + 1):
            if 0 <= s < n and 0 <= t < m:
                if abs(r - s) + abs(c - t) <= x:
                    total += grid[s][t]

    return total

sample2_grid = [
    [1, 3, 4, 1, 3, 1],
    [1, 1, 4, 1, 3, 1],
    [1, 1, 3, 2, 5, 3],
    [4, 3, 3, 1, 4, 1],
    [5, 2, 1, 1, 1, 1],
]

print(area_sum_box(sample2_grid, 2, 3))  # 預期 32
print(area_sum_box(sample2_grid, 0, 0))  # 預期 5
````

````python
def find_special_positions_box(grid):
    points = []
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            x = grid[r][c]
            total = area_sum_box(grid, r, c)
            if (total - x) % 10 == 0:
                points.append((r, c))
    return points

def solve_from_lines_box(lines):
    n, m = map(int, lines[0].split())
    grid = [list(map(int, lines[i].split())) for i in range(1, n + 1)]

    points = find_special_positions_box(grid)
    output = [str(len(points))]
    for r, c in points:
        output.append(f"{r} {c}")
    return "\n".join(output)

print(solve_from_lines_box(sample1))
print("---")
print(solve_from_lines_box(sample2))
````

## 11. 第一子題：`n = 1` 的一維情境

第一子題只有一列，所以座標永遠是 `(0, j)`。

這時候不需要算完整曼哈頓距離，只要看左右範圍：

- 左邊界：`max(0, j - x)`
- 右邊界：`min(m - 1, j + x)`

這一段可以先掌握第一子題解法，再推廣到完整二維情境。

````python
def solve_one_row(values):
    m = len(values)
    points = []

    for j in range(m):
        x = values[j]
        left = max(0, j - x)
        right = min(m - 1, j + x)
        total = sum(values[left:right + 1])

        if (total - x) % 10 == 0:
            points.append(j)

    output = [str(len(points))]
    for j in points:
        output.append(f"0 {j}")
    return "\n".join(output)

print(solve_one_row([3, 1, 4, 5, 1]))
````
