## 0. 題本解析

本課以 `APCS-程式實作中級題本範例.pdf` 的〈蒐集寶石〉為基礎。

題目給一個 `M x N` 的寶石地圖：

- `-1` 表示牆壁，機器人不能進入。
- `0` 表示沒有寶石，機器人可以進入，但走到這格時會停止。
- 正整數表示該格目前有幾顆寶石。

機器人狀態包含：

- 位置 `(r, c)`
- 方向，初始朝東方，也就是往右
- 得分 `score`，初始為 0
- 已取得寶石數 `gem_count`，初始為 0

每回合規則：

1. 如果目前格子沒有寶石，也就是 `grid[r][c] == 0`，停止。
2. 否則，把目前格子的寶石數加到 `score`，並取得一顆寶石。
3. 目前格子的寶石數減少 1，`gem_count += 1`。
4. 如果 `score` 是 `K` 的倍數，先右轉一次。
5. 如果前方是牆壁或界外，就持續右轉，直到前方可以走。
6. 前進一格，回到步驟 1。

## 0-1. 題目原文整理

### 問題描述

場地是一個 `M x N` 的二維陣列。每個元素 `(i, j)` 對應到第 `i` 橫列第 `j` 直行的格子，列由上而下、行由左而右，皆由 0 開始編號。

格子內容：

- `-1`：牆壁，機器人不能進入。
- `0`：沒有寶石，機器人可以進入，但走到後遊戲停止。
- 正整數：該格目前的寶石數量。

遊戲開始時：

- 機器人從 `(r, c)` 開始。
- 初始方向為東方，也就是往右。
- 初始得分 `score = 0`。
- 初始取得寶石數 `gem_count = 0`。

### 行動規則

每回合依序做以下事情：

1. 如果目前格子內無寶石，也就是 `grid[r][c] == 0`，停止遊戲。
2. 否則，將得分加上該格目前的寶石數量，並取得一顆寶石。該格剩餘寶石數減少一顆。
3. 如果總得分是 `K` 的倍數，向右轉，也就是順時針轉 90 度。
4. 若前方是牆壁或界外，則向右轉；持續右轉直到前方格子在範圍內且不是牆壁。
5. 進入前方格子，回到步驟 1。

注意：在原地右轉時，並不算再次走到該格子，也不會再次取得寶石。

### 輸入格式

第一行有五個整數：

```text
M N K r c
```

限制：

```text
1 <= M <= 100
2 <= N <= 100
2 <= K <= 20
```

接下來 `M` 行，每行有 `N` 個整數。`-1` 表示牆壁；否則為寶石數量，範圍是 `0` 到 `K-1`。

題目保證：

- 起始點不是牆壁。
- 起始點不會四周均無路可走。
- 機器人最終一定會停下來。

### 輸出格式

輸出一個整數，表示機器人停止時已取得的寶石數。

## 11. Debug trace：印出路徑

模擬題最容易錯在「順序」。除錯時可以先印出每次收寶石前的位置、收寶石前該格值、score、取得數、方向與下一步位置。

````python
def trace_simulation(grid, K, start_r, start_c, max_steps=20):
    r, c = start_r, start_c
    direction = 0
    score = 0
    gem_count = 0

    for step in range(1, max_steps + 1):
        if grid[r][c] == 0:
            print(step, (r, c), 'stop')
            break

        before = grid[r][c]
        score += before
        grid[r][c] -= 1
        gem_count += 1

        if score % K == 0:
            direction = turn_right_solution(direction)

        direction = rotate_until_can_move_solution(grid, r, c, direction)
        nr, nc = next_position_solution(r, c, direction)
        print(step, (r, c), 'before=', before, 'score=', score, 'count=', gem_count, 'dir=', direction, '->', (nr, nc))
        r, c = nr, nc

    return gem_count

sample2_grid = [list(map(int, line.split())) for line in sample2[1:]]
print(trace_simulation([row[:] for row in sample2_grid], 4, 2, 1, 10))
````

### 作業 D：整合版（不拆 function）
