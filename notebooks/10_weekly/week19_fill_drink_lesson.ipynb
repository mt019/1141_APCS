## 0. 題本解析

本課以 `APCS-程式實作初級題本範例.pdf` 的〈裝飲料〉為基礎。

杯子由下段與上段兩個長方體組成：

- 下段底面為 `w1 × w1`，高 `h1`，容積為 `w1 * w1 * h1`。
- 上段底面為 `w2 × w2`，高 `h2`，容積為 `w2 * w2 * h2`。

小夏依序倒入 N 杯水，每次倒水後水位上升若干高度。題目問的是 **所有次上升高度的最大值**。

注意：
- 杯子裝滿後水位不再上升，上升高度為 0。
- 題目保證每次上升高度皆為整數。

## 0-1. 題目原文整理

### 問題描述

杯子內層空間為上下兩個長方體：

- 下方：底面 `w1×w1 cm²`，高 `h1 cm`
- 上方：底面 `w2×w2 cm²`，高 `h2 cm`

依序倒入 N 杯冰水（體積單位 ml，1 cm³ = 1 ml），計算每杯倒入後水位上升的高度，輸出最大值。杯子滿後高度不再上升（上升值為 0）。

### 輸入格式

```text
N
w1 w2 h1 h2
V1 V2 ... VN
```

限制：`1 ≤ N ≤ 10`，`1 ≤ w1, w2, h1, h2 ≤ 50`，`w1 < w2`，每杯體積不超過 10⁶。

### 輸出格式

一個整數，為最大水位上升高度。

### 範例一


```text

```

## 8. Fix-it 1：高度計算沒有上限

下面程式少了「杯滿時高度為 `h1 + h2`」的條件，導致水量超過總容積時高度繼續增加。

````python
def height_from_volume_bug1(v, w1, h1, w2, h2):
    lv = w1 * w1 * h1
    if v <= lv:
        return v // (w1 * w1)
    else:
        return h1 + (v - lv) // (w2 * w2)  # BUG: 沒有限制最高為 h1 + h2

# total_cap = 4*4*8 + 6*6*5 = 128 + 180 = 308
print(height_from_volume_bug1(308, 4, 8, 6, 5))   # 應為 13，結果正確（剛好）
print(height_from_volume_bug1(2000, 4, 8, 6, 5))  # 應為 13，但結果會超出

# TODO: 寫出修正版
````

### Fix-it 1 解答

在 `if/elif/else` 最前面加上「滿杯」的判斷，避免水量超過容積時繼續計算。

````python
def height_from_volume_fixed1(v, w1, h1, w2, h2):
    lv = w1 * w1 * h1
    total_cap = lv + w2 * w2 * h2
    if v >= total_cap:
        return h1 + h2
    elif v <= lv:
        return v // (w1 * w1)
    else:
        return h1 + (v - lv) // (w2 * w2)

print(height_from_volume_fixed1(308, 4, 8, 6, 5))
print(height_from_volume_fixed1(2000, 4, 8, 6, 5))
````

## 9. Fix-it 2：把「目前高度」當成「上升高度」

下面程式把 `max_rise` 比較的是每次倒完水後的水位高度，而非這次上升的高度差。
當水位高度比上升高度大時（幾乎永遠如此），結果就會錯誤。

````python
def simulate_bug2(volumes, w1, h1, w2, h2):
    current_v = 0
    max_rise = 0
    total_cap = w1 * w1 * h1 + w2 * w2 * h2
    for cup_v in volumes:
        current_v = min(current_v + cup_v, total_cap)
        new_h = height_from_volume_solution(current_v, w1, h1, w2, h2)
        if new_h > max_rise:   # BUG: 比較的是水位高度，不是上升高度
            max_rise = new_h
    return max_rise

# 範例三正確答案是 23，但這個版本會輸出第三杯後的水位高度 42
print(simulate_bug2([400, 600, 2500, 3000], 5, 12, 10, 30))

# TODO: 寫出修正版
````

### Fix-it 2 解答

要比較的是「倒這杯水前後的高度差」，而不是「倒完後的高度」。需要在倒水前先記住舊高度。

````python
def simulate_fixed2(volumes, w1, h1, w2, h2):
    current_v = 0
    max_rise = 0
    total_cap = w1 * w1 * h1 + w2 * w2 * h2
    for cup_v in volumes:
        old_h = height_from_volume_solution(current_v, w1, h1, w2, h2)
        current_v = min(current_v + cup_v, total_cap)
        new_h = height_from_volume_solution(current_v, w1, h1, w2, h2)
        rise = new_h - old_h
        if rise > max_rise:
            max_rise = rise
    return max_rise

print(simulate_fixed2([400, 600, 2500, 3000], 5, 12, 10, 30))
````

## 11. Debug trace：印出每杯的詳細狀態

水位計算容易出錯在分段條件或上限。除錯時逐杯印出舊水量、新水量、舊高度、新高度、上升值。

````python
def trace_simulation(volumes, w1, h1, w2, h2):
    current_v = 0
    total_cap = w1 * w1 * h1 + w2 * w2 * h2
    for i, cup_v in enumerate(volumes, 1):
        old_h = height_from_volume_solution(current_v, w1, h1, w2, h2)
        new_v = min(current_v + cup_v, total_cap)
        new_h = height_from_volume_solution(new_v, w1, h1, w2, h2)
        rise = new_h - old_h
        print(f"第 {i} 杯：倒入 {cup_v}ml，水量 {current_v}→{new_v}，高度 {old_h}→{new_h}，上升 {rise}")
        current_v = new_v

trace_simulation([400, 600, 2500, 3000], 5, 12, 10, 30)
````
