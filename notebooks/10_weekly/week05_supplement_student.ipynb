## 1. Fact 與 Statement 的差別

在程式裡，布林變數通常代表一個 Statement，也就是「可以回答是或不是」的句子。Fact 是實際發生的情況，我們用 Fact 來判斷 Statement 是 `True` 還是 `False`。


- Fact：觀察到的事情，例如「我看到外面在下雨」。
- Statement：程式裡要判斷的句子，例如 `A = 今天正在下雨。`
- 當 Statement 對應的 Fact 成立時，該布林變數就是 `True`，反之為 `False`。


### 課堂中的示例

| 代碼 | Statement（句子） | 如果為 True | 如果為 False |
| --- | --- | --- | --- |
| A | 今天正在下雨。 | 出門看到雨滴或地面濕。 | 天氣晴朗或只有陰天。 |
| B | 我有帶傘出門。 | 在包包裡拿出雨傘。 | 忘記帶傘或傘不在身邊。 |

閱讀下方輸出，觀察四種組合代表的情境，理解 `A and B` 與 `A or B` 的差別。


```python
scenarios = [
    {"A": False, "B": False},
    {"A": True, "B": False},
    {"A": False, "B": True},
    {"A": True, "B": True},
]

def describe_scenario(a_is_true, b_is_true):
    rain_fact = "正在下雨" if a_is_true else "今天沒有下雨"
    umbrella_fact = "我帶著雨傘" if b_is_true else "我沒有帶傘"
    return rain_fact, umbrella_fact

for item in scenarios:
    rain, umbrella = describe_scenario(item["A"], item["B"])
    result_and = item["A"] and item["B"]
    result_or = item["A"] or item["B"]
    print(f"A={item['A']:5}  B={item['B']:5}  ｜ {rain}，{umbrella}")
    print(f"    A and B → {result_and}    A or B → {result_or}\n")
```

## 2. 迴圈控制關鍵字快速整理

下表整理了幾個常見的迴圈控制方式。讀完後請看範例輸出，想一想流程為什麼會改變。


| 關鍵字 / 模式 | 作用 | 何時使用 | 小提醒 |
| --- | --- | --- | --- |
| `break` | 立刻跳出目前這一層迴圈。 | 找到答案後就不用再繼續找。 | `break` 後面的程式不會執行。 |
| `continue` | 略過本輪剩下的程式，直接進下一輪。 | 想跳過不需要處理的資料。 | 記得在 `continue` 之前完成必要的更新。 |
| `pass` | 什麼都不做，只有佔位的效果。 | 先把架構寫好，日後再補內容。 | `pass` 不會自動跳出或繼續，要自己控制流程。 |
| `while ... else` / `for ... else` | 如果整個迴圈從頭到尾都沒有 `break`，才會執行 `else`。 | 搜尋所有資料後仍沒找到時給提示。 | 只要用到 `break`，`else` 就不會執行。 |


```python
numbers = [5, 3, -1, 2, 0, 4]

print("使用 break：遇到第一個負數就停止")
for n in numbers:
    if n < 0:
        print(f"遇到 {n} ➜ break")
        break
    print(f"處理 {n}")
else:
    print("for-else：沒有 break 時才執行")

print("\n使用 continue：負數不處理，但迴圈仍繼續")
for n in numbers:
    if n < 0:
        print(f"跳過 {n}")
        continue
    print(f"保留 {n}")

print("\n使用 pass：先佔位，之後再補邏輯")
for n in numbers:
    if n == 0:
        pass  # 之後想好要怎麼處理 0 再補上
    else:
        print(f"處理 {n}")
```

## 3. 密碼錯誤次數限制的多種寫法

我們要練習把「密碼最多輸入幾次」這個需求寫成程式，並避免出現多給一次機會的錯誤。先看一個穩定的範例：


```python
def make_fake_input(answers):
    """建立模擬 input() 的函式，依序回傳準備好的答案。"""
    answers = list(answers)
    def fake_input(prompt):
        if not answers:
            raise RuntimeError("沒有更多輸入，請再提供答案清單。")
        value = answers.pop(0)
        print(f"{prompt}{value}")
        return value
    return fake_input
```

```python
def login_with_break(mock_input, password="apcs", max_tries=3):
    """示範：使用 while + break 控制登入流程。"""
    tries = 0
    while tries < max_tries:
        user_input = mock_input(f"第 {tries + 1} 次輸入：")
        if user_input == password:
            print("登入成功！")
            return True
        tries += 1
        remaining = max_tries - tries
        if remaining:
            print(f"還剩 {remaining} 次機會。\n")
        else:
            print("沒有剩餘次數了。\n")
    print("登入失敗，帳號暫時鎖定。")
    return False
```

### 測試看看

執行下方範例，觀察成功與失敗時輸出的順序。你也可以修改答案清單，模擬不同的情況。


```python
print("--- 範例一：第二次成功 ---")
login_with_break(
    make_fake_input(["0000", "apcs"]),
    password="apcs",
    max_tries=3,
)

print("\n--- 範例二：全部失敗 ---")
login_with_break(
    make_fake_input(["1234", "8888", "9999"]),
    password="apcs",
    max_tries=3,
)
```

### 常見錯誤：多給一次機會

下面的程式會在 `while` 迴圈之前先讀一次輸入，但沒有把這次輸入算進 `tries`，結果就會讓使用者多輸入一次。執行看看，注意「剩餘次數」的提示是否合理。


```python
def login_off_by_one(mock_input, password="apcs", max_tries=3):
    tries = 0
    user_input = mock_input("請輸入密碼：")
    while user_input != password and tries != max_tries:
        tries += 1
        remaining = max_tries - tries
        print(f"第 {tries} 次錯誤，剩餘 {remaining} 次。")
        user_input = mock_input("再試一次：")
    if tries == max_tries:
        print("登入失敗（注意：這版會多給一次機會）。")
        return False
    print("登入成功（但要小心這版的次數計算）。")
    return True

print("--- 範例：觀察多給一次的情況 ---")
login_off_by_one(
    make_fake_input(["0000", "1111", "2222", "apcs"]),
    password="apcs",
    max_tries=3,
)
```
