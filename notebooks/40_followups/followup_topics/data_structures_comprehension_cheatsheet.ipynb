## Tuple 語法速查
| 操作 | 範例 | 說明 |
| --- | --- | --- |
| 建立 | `pair = (name, score)` | 逗號決定 tuple，括號可省略。|
| 拆封 | `name, score = pair` | 元素數量需匹配。|
| 巢狀拆封 | `name, (math, eng) = record` | 可對內層 tuple/list 再拆。|
| 取值 | `pair[0]` | 讀取但不可修改。|
| 與 enumerate | `for idx, value in enumerate(data):` | 在 trace APCS 題目時常用。|

````python
# Tuple 拆封示例：APCS 程式識讀題目 2 的 trace 列表
log = [
    (0, 1, '[1]'),
    (0, 2, '[2]'),
    (1, 3, '[3]'),
]
for round_idx, (i_value, j_value, output) in enumerate(log, start=1):
    print(f"Round {round_idx}: i={i_value}, j={j_value}, print {output}")
````

## List 語法速查
| 操作 | 範例 | 說明 |
| --- | --- | --- |
| 建立 | `scores = [78, 85, 90]` | 常用於輸入暫存、trace。|
| 附加 | `scores.append(95)` | 加入單一元素。|
| 延伸 | `scores.extend([60, 88])` | 加入另一個 iterable。|
| 切片 | `scores[1:3]` | 取得區間，`[:3]` 與 `[3:]` 代表從頭/到尾。|
| 推導 | `[score+5 for score in scores]` | 生成新 list。|
| 排序 | `sorted(scores, reverse=True)` | 回傳新串列；`scores.sort()` 就地修改。|

````python
# List 操作示例：七言對聯輸入
line1 = [1, 0, 0, 0, 0, 1, 1]
line2 = [0, 1, 0, 0, 1, 1, 0]
# 比較第 2, 4, 6 個元素
for idx in (1, 3, 5):
    print(f"位置 {idx+1}: line1={line1[idx]}, line2={line2[idx]}")
````

## Dict 語法速查
| 操作 | 範例 | 說明 |
| --- | --- | --- |
| 建立 | `stats = {'Amy': 2}` | key 必須可雜湊。|
| 讀取含預設 | `stats.get(name, 0)` | 避免 `KeyError`。|
| 更新 | `stats[name] = stats.get(name, 0) + 1` | APCS 計數常見。|
| `setdefault` | `tags.setdefault(name, set()).add(tag)` | 先建立預設容器。|
| 遍歷 | `for name, count in stats.items():` | 同時取得 key/value。|
| 合併 | `stats     \|= new_stats` (3.9+) | 合併並覆寫重複 key。|

````python
# Dict 操作示例：summarize_attempts 精簡版
logs = [
    ('Amy', True),
    ('Bob', False),
    ('Amy', False),
]
stats = {}
success = set()
for name, ok in logs:
    stats[name] = stats.get(name, 0) + 1
    if ok:
        success.add(name)
print('stats =', stats)
print('success =', success)
````

## Set 語法速查
| 操作 | 範例 | 說明 |
| --- | --- | --- |
| 建立 | `focus = {'dict', 'set'}` 或 `set()` | 空集合需用 `set()`。|
| 加入 | `focus.add('tuple')` | 若已存在不重複。|
| 聯集 | `done  \| todo` | 合併主題。|
| 交集 | `done & target` | 找出已完成且當前需要的。|
| 差集 | `todo - done` | 尚未完成項目。|
| 判斷 | `subset <= superset` | 是否全部完成。|

````python
# Set 操作示例：進度對照
roadmap = {'tuple', 'dict', 'set', 'comprehension'}
today_focus = {'range', 'tuple', 'comprehension'}
need_review = today_focus - roadmap
print('need_review =', need_review)
print('已在計畫內的項目 =', today_focus & roadmap)
````

## Comprehension 對照表（正常寫法 vs 簡寫）
> Comprehension 指的是用一行語法快速建立 list / dict / set 的語法糖：`[expression for ... if ...]`、`{key: value for ...}`、`{expr for ...}`。
> 特色：將「迴圈 + 條件 + append/assign」壓縮成一行，適合在資料轉換、篩選時讓程式更精簡；不過當轉換邏輯過長或需要多步調試時，仍建議先寫成一般迴圈再逐步濃縮。
> 步驟：
> 1) 先用正常寫法完成並確認輸出 → 
> 2) 找出 expression / for / if 對應 → 
> 3) 轉成 comprehension，記得保留可讀性與註解。


### List comprehension
#### 範例 1：平均成績
正常寫法：
```python
avg = []
for name, scores in students:
    average = sum(scores) / len(scores)
    avg.append((name, round(average, 2)))
```
comprehension 寫法：
```python
avg = [
    (name, round(sum(scores)/len(scores), 2))
    for name, scores in students
]
```
#### 範例 2：篩選成功者
正常寫法：
```python
success = []
for name, ok in logs:
    if ok:
        success.append(name)
```
comprehension 寫法：
```python
success = [name for name, ok in logs if ok]
```


````python
students = [
    ('Amy', [92, 88, 95]),
    ('Ben', [75, 81, 79]),
    ('Chloe', [84, 90, 87]),
]
logs = [
    ('Amy', True),
    ('Ben', False),
    ('Chloe', True),
]
avg_loop = []
for name, scores in students:
    average = sum(scores) / len(scores)
    avg_loop.append((name, round(average, 2)))
avg_comp = [(name, round(sum(scores)/len(scores), 2)) for name, scores in students]
success_loop = []
for name, ok in logs:
    if ok:
        success_loop.append(name)
success_comp = [name for name, ok in logs if ok]
print('avg_loop  =', avg_loop)
print('avg_comp  =', avg_comp)
print('success_loop =', success_loop)
print('success_comp =', success_comp)
assert avg_loop == avg_comp
assert success_loop == success_comp
print('List comprehension 寫法與正常寫法輸出一致 ✅')
````

### Dict comprehension
#### 範例 1：字母對應平方
正常寫法：
```python
result = {}
for ch in 'ABC':
    result[ch] = ord(ch) ** 2
```
comprehension 寫法：
```python
result = {ch: ord(ch) ** 2 for ch in 'ABC'}
```
#### 範例 2：過濾計數
正常寫法：
```python
filtered = {}
for k, v in stats.items():
    if v >= 2:
        filtered[k] = v
```
comprehension 寫法：
```python
filtered = {k: v for k, v in stats.items() if v >= 2}
```


````python
stats = {'Amy': 3, 'Ben': 1, 'Chloe': 4}
result_loop = {}
for ch in 'ABC':
    result_loop[ch] = ord(ch) ** 2
result_comp = {ch: ord(ch) ** 2 for ch in 'ABC'}
filtered_loop = {}
for k, v in stats.items():
    if v >= 2:
        filtered_loop[k] = v
filtered_comp = {k: v for k, v in stats.items() if v >= 2}
print('result_loop =', result_loop)
print('result_comp =', result_comp)
print('filtered_loop =', filtered_loop)
print('filtered_comp =', filtered_comp)
assert result_loop == result_comp
assert filtered_loop == filtered_comp
print('Dict comprehension 寫法與正常寫法輸出一致 ✅')
````

### Set comprehension
#### 範例 1：蒐集完成題型
正常寫法：
```python
completed = set()
for topic in today_focus:
    if topic in done:
        completed.add(topic)
```
comprehension 寫法：
```python
completed = {topic for topic in today_focus if topic in done}
```
#### 範例 2：偶數集合
正常寫法：
```python
evens = set()
for i in range(10):
    if i % 2 == 0:
        evens.add(i)
```
comprehension 寫法：
```python
evens = {i for i in range(10) if i % 2 == 0}
```


````python
done = {'tuple', 'dict', 'range'}
today_focus = {'tuple', 'dict', 'comprehension'}
completed_loop = set()
for topic in today_focus:
    if topic in done:
        completed_loop.add(topic)
completed_comp = {topic for topic in today_focus if topic in done}
evens_loop = set()
for i in range(10):
    if i % 2 == 0:
        evens_loop.add(i)
evens_comp = {i for i in range(10) if i % 2 == 0}
print('completed_loop =', completed_loop)
print('completed_comp =', completed_comp)
print('evens_loop =', sorted(evens_loop))
print('evens_comp =', sorted(evens_comp))
assert completed_loop == completed_comp
assert evens_loop == evens_comp
print('Set comprehension 寫法與正常寫法輸出一致 ✅')
````
