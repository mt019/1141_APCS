## 2. 基礎 list 模擬
以下範例透過 `list` + `pop(0)` 模擬排隊，請逐行觀察輸出，理解 head/tail 的移動。


```python
queue = []
for person in ['Amy', 'Ben', 'Clara']:
    queue.append(person)
    print('加入: ', person, '->', queue)

print('服務: ', queue.pop(0))
print('剩餘: ', queue)
print('服務: ', queue.pop(0))
print('剩餘: ', queue)

queue.append('David')
queue.append('Ella')
print('加入兩人後: ', queue)
print('服務: ', queue.pop(0))
print('剩餘: ', queue)
```

## 3. 使用 collections.deque
`deque` 支援 O(1) 的 append 與 popleft，非常適合佇列情境。


```python
from collections import deque
queue = deque(['Amy', 'Ben', 'Clara'])
print(queue)
queue.append('David')
print('加入 David ->', queue)
served = queue.popleft()
print('服務:', served, '| 剩餘:', queue)
```
