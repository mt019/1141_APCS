## 2. 星號圖形輸出
練習使用巢狀迴圈與格式化控制空白與星號：


```python
def print_triangle(rows):
    for i in range(rows):
        spaces = ' ' * (rows - i - 1)
        stars = '*' * (2 * i + 1)
        print(spaces + stars)

print_triangle(4)
```
