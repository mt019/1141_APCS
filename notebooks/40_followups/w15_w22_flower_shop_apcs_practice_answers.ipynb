## 題目 2：花店商品資料結構（解答）


```python
# 參考解答

def total_value(items):
    return sum(item["price"] * item["stock"] for item in items)

items = [
    {"name": "rose", "price": 30, "stock": 12},
    {"name": "lily", "price": 25, "stock": 8},
    {"name": "tulip", "price": 28, "stock": 5},
]

print(total_value(items))
```

## 題目 3：NamedTuple vs dataclass（解答）


```python
# 參考解答
from typing import NamedTuple
from dataclasses import dataclass

class FlowerNT(NamedTuple):
    name: str
    price: int
    stock: int

@dataclass
class FlowerDC:
    name: str
    price: int
    stock: int

f1 = FlowerNT("rose", 30, 12)
# f1.stock += 5  # 這行應該會錯

f2 = FlowerDC("lily", 25, 8)
f2.stock += 5

print(f1)
print(f2)
```

## 題目 6：CSV 讀取與統計（解答）


```python
# 參考解答
import io, csv

csv_text = """name,price,stock
rose,30,12
lily,25,8
tulip,28,5
"""

f = io.StringIO(csv_text)
reader = csv.DictReader(f)

total_value = 0
for row in reader:
    total_value += int(row["price"]) * int(row["stock"])

print(total_value)
```

## 題目 8：APCS 輸入格式（固定筆數）（解答）


```python
# 參考解答
from typing import List, Dict

def parse_fixed(data: str) -> List[Dict[str, int]]:
    lines = [line.strip() for line in data.strip().split("\n") if line.strip()]
    n = int(lines[0])
    items = []
    for i in range(1, 1 + n):
        name, price, stock = lines[i].split()
        items.append({"name": name, "price": int(price), "stock": int(stock)})
    return items

sample = """3
rose 30 12
lily 25 8
tulip 28 5
"""

print(parse_fixed(sample))
```

## 題目 9：APCS 輸入格式（讀到 EOF）（解答）


```python
# 參考解答
from typing import List, Dict

def parse_until_eof(data: str) -> List[Dict[str, int]]:
    items = []
    for line in data.strip().split("\n"):
        line = line.strip()
        if not line:
            continue
        name, price, stock = line.split()
        items.append({"name": name, "price": int(price), "stock": int(stock)})
    return items

sample = """rose 30 12
lily 25 8
tulip 28 5
"""

print(parse_until_eof(sample))
```

## 題目 11：錯誤處理（輸入驗證）（解答）


```python
# 參考解答

def safe_read_order(line: str):
    parts = line.split()
    if len(parts) != 3:
        return None
    name, price_s, qty_s = parts
    if not price_s.isdigit() or not qty_s.isdigit():
        raise ValueError("price/qty must be positive integers")
    price = int(price_s)
    qty = int(qty_s)
    return {"name": name, "price": price, "qty": qty}

samples = [
    "rose 30 2",
    "lily -5 1",
    "tulip 28",
    "sunflower 40 3"
]

for s in samples:
    try:
        print(safe_read_order(s))
    except ValueError as e:
        print("ValueError:", e)
```

## 題目 12：綜合題（花店訂單 + APCS 風格輸入）（解答）


```python
# 參考解答

def solve(data: str) -> str:
    lines = [line.strip() for line in data.splitlines() if line.strip()]
    m = int(lines[0])
    stats = {}
    for i in range(1, 1 + m):
        name, price_s, qty_s = lines[i].split()
        price = int(price_s)
        qty = int(qty_s)
        if name not in stats:
            stats[name] = {"qty": 0, "income": 0}
        stats[name]["qty"] += qty
        stats[name]["income"] += price * qty

    out_lines = []
    for name in sorted(stats.keys()):
        out_lines.append(f"{name} {stats[name]['qty']} {stats[name]['income']}")
    return " ".join(out_lines)

sample = """5
rose 30 2
lily 25 3
rose 30 1
tulip 28 2
lily 25 1
"""

print(solve(sample))
```
