## 語法鋪墊：tuple 與 list 重點
- **tuple**：不可變，用來打包固定欄位；支援拆封、巢狀結構。
- **list**：可變序列，可追加、刪除、切片，常配合迴圈與推導式。
- **拆封技巧**：`a, b = pair`、`first, *rest = items`、`*_, last = items`，搭配 `for name, scores in data` 讓程式更好讀。
- **列表方法**：`append`、`extend`、`pop`、`remove`、`index` 等，用於管理動態資料。


```python
# 範例：tuple 與 list 的基本操作
point = (25.0478, 121.5319)
lat, lng = point
print('lat =', lat, '| lng =', lng)

scores = [82, 91, 88]
scores.append(95)
front = scores[:2]
tail = scores[-2:]
print('scores =', scores)
print('front =', front, '| tail =', tail)

students = [('Amy', [82, 91, 88]), ('Ben', [70, 76, 80])]
for name, records in students:
    avg = round(sum(records) / len(records), 1)
    print(f'{name} -> 平均 {avg}')
```

## 語法鋪墊：dict 關鍵語法
- `get(key, default)`：讀取時附帶預設值，避免 KeyError。
- `setdefault(key, default)`：若不存在就建立並回傳該值。
- `update`：批次合併或覆寫鍵值。
- `items()` / `values()` / `keys()`：常用於迴圈遍歷或轉換。
- dict comprehension：`{key: expr for key, value in data}`，快速建立新字典。


```python
# 範例：dict 常見操作
profile = {'name': 'Amy', 'tries': 2}
print('name ->', profile.get('name'))
print('last_login ->', profile.get('last_login', '尚未登入'))

profile.setdefault('tags', set()).add('math')
profile.setdefault('tags', set()).add('python')
print('tags ->', profile['tags'])

profile.update({'tries': profile['tries'] + 1, 'level': 'bronze'})
print('更新後 profile ->', profile)

length_map = {name: len(name) for name in ['Amy', 'Ben', 'Cindy']}
print('length_map ->', length_map)
```
