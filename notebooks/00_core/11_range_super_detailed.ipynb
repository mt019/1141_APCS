這份筆記系統整理 Python 內建的 `range()` 函式/型別，從最基本語法到實務範例。
使用 `range` 時，請記得它不會一次把所有數字存進記憶體，而是按需生成 (lazy evaluation)。


## 基本語法總覽

- `range(stop)`: 從 0 數到 `stop-1`。
- `range(start, stop)`: 從 `start` 數到 `stop-1`。
- `range(start, stop, step)`: 每次增加 (或減少) `step`，直到超過邊界。`step` 不能為 0。

`range` 的 `stop` 參數永遠 **不包含** 在序列中，是半開區間 `[start, stop)`。


```python
print(list(range(10)))
```

```python
print(list(range(5,10)))
```

```python
print(list(range(1,11)))
```

```python
print(list(range(1,12,2)))
```

```python
print(list(range(10,-1,-1)))
```

```python
# range 物件本身是可迭代物件，但不是 list
r = range(5)
print(r)
print(type(r))
print(list(r))  # 轉成 list 查看內容
```

## range 與記憶體效率

- `range` 是不可變序列 (immutable sequence)，類似於 tuple，支援 `len()`、索引與成員檢查。
- 它不會預先建立完整清單，所以即使 `range(0, 1_000_000_000)` 也只佔用很小的記憶體。
- 需要真實資料時再用 `list(range(...))` 或 `tuple(range(...))` 轉換。


```python
r = range(2, 11, 2)
print('長度 =', len(r))
print('索引 0 =', r[0])
print('索引 1 =', r[1])
print('5 是否在 range 中? ->', 5 in r)
print('10 是否在 range 中? ->', 10 in r)
```

## start、stop、step 舉例

`start` 預設為 0，`step` 預設為 1。下面例子顯示不同參數的差異。


```python
print('range(5) ->', list(range(5)))
print('range(2, 7) ->', list(range(2, 7)))
print('range(2, 10, 2) ->', list(range(2, 10, 2)))
```

## 正向遞增 step

正的 `step` 代表往右 (由小到大) 生成數字，當下一個數字 >= `stop` 時停止。


```python
for number in range(3, 12, 3):
    print(number)
```

## 反向遞減 step

當 `step` 為負數時，`start` 必須大於 `stop`，序列才會往左生成。
常見於需要倒數或反向走訪時。


```python
print(list(range(10, 0, -2)))
for countdown in range(5, 0, -1):
    print('倒數:', countdown)
```

## `step` 不能為 0

如果嘗試讓 `step` 為 0，Python 會丟出 `ValueError`，提醒你無法前進或後退。


```python
try:
    range(1, 5, 0)
except ValueError as exc:
    print('捕捉到錯誤:', exc)
```

## range 與 for 迴圈

`for` 迴圈會逐一向 `range` 要下一個整數，直到耗盡序列。這種配合最常見於重複執行固定次數的工作。


```python
for i in range(4):
    print(f'i = {i}')
```

## 常見進階用法

- `sum(range(n))`：快速取得等差級數總和。
- `enumerate(iterable)` 搭配 `range` 保留索引。
- `range` 可用於多重迴圈、建立座標、批次處理資料等情境。
- 可以透過 `range(start, stop, step)` 迭代偶數、奇數或任何固定間隔的序列。


```python
print('sum(range(10,-1,-1)) =', sum(range(10,-1,-1)))
```

```python
print('sum(range(5)) =', sum(range(5)))
items = ['apple', 'banana', 'citrus']
for idx in range(len(items)):
    print(idx, items[idx])

for row in range(1, 4):
    for col in range(1, 3):
        print(f'(row={row}, col={col})', end=' ')
    print()
```

## 容易犯的錯

- 忘記 `stop` 不包含在序列內，造成少算或多算一次。
- 在需要浮點間隔時直接用 `range`（它只支援整數），這種情境通常改用 `numpy.arange` 或自己累加。
- 忽略 `step` 正負號，導致空序列。


## 綜合範例：倒數並計算加總

以下示範如何使用 `range(10, 0, -1)` 做倒數，同時把每次的數值累加起來。


```python
total = 0
for i in range(10, 0, -1):
    total += i
    print(i)

print('total =', total)
```
