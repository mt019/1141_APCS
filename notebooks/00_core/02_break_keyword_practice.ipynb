本筆記示範如何在 `for` 與 `while` 迴圈中使用 `break`，以及觀察程式在提早結束時的行為。

## 1. 在 `for` 迴圈中找到第一個符合條件的值
我們想找出第一個大於 10 的數字，找到後就停止搜尋。

```python
numbers = [3, 8, 5, 11, 4, 15]
first_over_ten = None
steps = 0

for value in numbers:
    steps += 1
    if value > 10:
        first_over_ten = value
        break  # 找到符合條件的值，提早結束迴圈

print('第一個大於 10 的數字:', first_over_ten)
print('檢查元素數量:', steps)
```

> 任務：將 `numbers` 換成不同的排列，觀察 `steps` 是否跟著改變。

## 2. `for` 迴圈搭配 `else`
`for` 後面的 `else` 只會在迴圈沒有被 `break` 中斷時執行。

```python
targets = [7, 9, 12]
numbers = [1, 4, 6, 8]

for target in targets:
    if target in numbers:
        print(target, '在列表裡，提早結束搜尋。')
        break
else:
    print('沒有找到任何目標值。')
```

> 任務：試著把 `numbers` 改成 `[1, 4, 6, 8, 12]`，思考 `else` 會不會執行。

## 3. `while` 迴圈控制次數
使用 `while` 迴圈時，可以搭配 `break` 避免無窮迴圈。

```python
count = 0
limit = 5

while True:  # 先假設要一直重複
    count += 1
    print('目前次數:', count)
    if count == limit:
        break  # 達到限制就跳出

print('迴圈已結束。')
```

> 思考：若忘記 `break`，這段程式會發生什麼事？如何避免？
