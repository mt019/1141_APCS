## 1) 單行基本輸入

### 1-1 讀取單一整數


````python
# 範例輸入：
# 42

n = int(input())
print(n)
````

### 1-2 同一行多個整數（最常用）

題目常見：`n m`、`a b`、`x y z`。


````python
# 範例輸入：
# 3 9

a, b = map(int, input().split())
print(a, b, a + b)
````

### 1-3 讀一整行，轉成 list

適合：
- 一行有很多數字
- 題目說「接下來一行有 k 個數」但你不想硬寫 k


````python
# 範例輸入：
# 10 20 30 40

data = list(map(int, input().split()))
print(data)
print(sum(data))
````

## 2) 多行固定筆數輸入（先給 n，再讀 n 行）

題型：
- 第一行給 `n`
- 接下來有 `n` 行資料（每行格式固定）

最通用模板：

```python
n = int(input())
for _ in range(n):
    row = list(map(int, input().split()))
    # 處理 row
```


````python
# 範例輸入：
# 3
# 1 2
# 3 4
# 5 6

n = int(input())
pairs = []
for _ in range(n):
    a, b = map(int, input().split())
    pairs.append((a, b))

print(pairs)
````

## 3) 讀到 EOF（ZeroJudge 最常見陷阱）

很多題目不會告訴你總共有幾筆資料，只說：
- 「讀到檔案結束」

這時你有兩種主流寫法：

### 3-A) `try/except EOFError`（最直覺）

```python
while True:
    try:
        line = input()
    except EOFError:
        break
    # 處理 line
```

### 3-B) 直接迭代 `sys.stdin`（通常更乾淨）

```python
import sys
for line in sys.stdin:
    data = line.split()
    if not data:
        continue
```

> 注意：有些測資會包含空行，建議先 `split()` 後檢查。

補充：在 Jupyter Notebook 中 `input()` 不能用 Ctrl-D 送出 EOF，
因此可改用「空行結束」作為示範。


### 3-A-Notebook) 空行結束（Notebook 示範用）

在 Notebook 中可用空行代表輸入結束：


````python
nums = []
while True:
    line = input()
    if line == '':
        break
    nums.append(int(line))

print(nums, 'sum =', sum(nums))
````

````python
# 範例輸入（檔案結束 EOF 後停止）：
# 10
# 20
# 30

nums = []
while True:
    try:
        line = input()
    except EOFError:
        break
    if not line.strip():
        continue
    nums.append(int(line))

print(nums, 'sum =', sum(nums))
````

````python
# 範例輸入（含空行；檔案結束 EOF 後停止）：
# 1 2
#
# 3 4

pairs = []
while True:
    try:
        line = input()
    except EOFError:
        break
    data = line.split()
    if not data:
        continue
    a, b = map(int, data)
    pairs.append((a, b))

print(pairs)
````

## 4) 高效輸入：`sys.stdin.readline` vs `sys.stdin.read`

當資料量大時（例如超大矩陣、很多筆查詢），`input()` 可能 TLE。

### 4-1 用 `sys.stdin.readline` 取代 `input()`

```python
import sys
input = sys.stdin.readline

n, m = map(int, input().split())
```

### 4-2 一次讀完：`sys.stdin.read().split()`

這招很快，但寫法會更「指標化」：

```python
import sys
it = iter(sys.stdin.read().split())

n = int(next(it))
# 接著一直 next(it)
```

優點：快  
缺點：程式可讀性較差、記憶體用量較高


````python
# 這段是「大量資料」時的加速模板（概念示範用）。
# 在 Notebook 不建議搭配前面的示範輸入。
#
# import sys
# it = iter(sys.stdin.read().split())
# n = int(next(it))
# for _ in range(n):
#     a = int(next(it)); b = int(next(it))
#     print(a + b)

print('請看上方註解模板（大量資料用）')
````

## 6) 小練習：直接用題目樣式讀入（APCS 骰子範例的輸入部分）

以題目常見格式：
- 第一行 `n m`
- 接著 `m` 行 `a b`

目標：只做「讀進來並印出來」，確認你的輸入解析沒有問題。


````python
# 範例輸入：
# 3 3
# 2 -1
# 3 -2
# 3 1

n, m = map(int, input().split())
ops = [tuple(map(int, input().split())) for _ in range(m)]

print('n =', n, 'm =', m)
print('ops =', ops)
````

## 7) 改錯練習（Fix-it）：輸入解析你真的會嗎？

說明：下面每題都給一段「常見的錯誤輸入程式」。

你的任務：
1) 先說明錯在哪（會出現什麼錯誤/為什麼）
2) 再把它改對

> 這些錯誤在 ZeroJudge 最常見，因為測資可能有空行、可能讀到 EOF、可能每行數量不固定。


### Fix-it 1：多讀一行 / 少讀一行（讀法與題意不一致）


````python
# 題目格式：第一行 n，接著 n 行，每行兩個整數。
# 問題：下面程式多讀了一行，導致後面整個錯位。請修正。

n = int(input())
input()  # ❌ 多讀了一行（把第一筆資料吃掉了）

pairs = []
for _ in range(n):
    a, b = map(int, input().split())
    pairs.append((a, b))

print(pairs)
````

修正方向：刪掉那個多餘的 `input()`，並用範例輸入驗證。


### Fix-it 2：空行導致 unpack 爆炸


````python
# 問題：如果測資中間有空行，這段會 ValueError。
# 請修正：遇到空行要略過。

while True:
    try:
        a, b = map(int, input().split())
    except EOFError:
        break

    print(a + b)
````

````python
# 一種修正範例
while True:
    try:
        line = input()
    except EOFError:
        break

    data = line.split()
    if not data:
        continue

    a, b = map(int, data)
    print(a + b)
````

### Fix-it 3：每行數量不固定卻用 a,b 解包


````python
# 問題：有些題每行可能是 1 個數或很多個數。
# 這段硬解包 a,b 會直接炸。請改成「先讀成 list 再處理」。

a, b = map(int, input().split())
print(a, b)
````

````python
# 一種修正範例
nums = list(map(int, input().split()))
print('count =', len(nums))
print(nums)
````

### Fix-it 4：EOF 題卻用固定筆數讀（會卡住等待輸入）


````python
# 題目其實是「讀到 EOF」，但你用固定筆數去讀，會卡住或讀不到完整資料。
# 請修正成 EOF 模式。

n = int(input())  # ❌ 這題根本沒有給 n
for _ in range(n):
    x = int(input())
    print(x)
````

修正方向：用 `while True + try/except EOFError` 或 `for line in sys.stdin`（上站時才用）。


## 8) 最常見錯誤與快速排查

1) **`ValueError: not enough values to unpack`**
- 你寫 `a, b = ...` 但那行其實不是兩個數
- 先 `print(line)` 或先 `data = input().split(); print(data)` 看看

2) **讀到空行**
- 有些測資會有空白行
- `data = line.split()` 後，`if not data: continue`

3) **用 `input()` 太慢 TLE**
- 改 `input = sys.stdin.readline`
- 或用 `sys.stdin.read().split()`

4) **iterator 只能走一次**
- `map(...)` / `zip(...)` 都是 iterator
- 要重複用就轉 `list(...)`
