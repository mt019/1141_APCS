## 1. 何時用 CSV？何時用 JSON？
- **CSV**：表格型資料（每列同樣欄位）
- **JSON**：巢狀資料（物件、清單、巢狀結構）

## 2. CSV 基礎：寫入與讀取

```python
import csv

rows = [
    {"name": "Amy", "score": 88},
    {"name": "Ben", "score": 92},
]

with open("scores.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "score"])
    writer.writeheader()
    writer.writerows(rows)
```

```python
import csv

loaded = []
with open("scores.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        loaded.append({"name": row["name"], "score": int(row["score"])})

loaded
```

## 3. CSV 小例子：簡易成績統計

```python
avg = sum(r["score"] for r in loaded) / len(loaded)
avg
```

## 4. CSV 追加寫入（append）

```python
import csv

with open("scores.csv", "a", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "score"])
    writer.writerow({"name": "Cindy", "score": 95})
```

## 5. JSON 基礎：寫入與讀取

```python
import json

data = {
    "class": "10A",
    "students": [
        {"name": "Amy", "score": 88},
        {"name": "Ben", "score": 92},
    ]
}

with open("class.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
import json

with open("class.json", "r", encoding="utf-8") as f:
    data_loaded = json.load(f)

data_loaded
```

## 6. JSON 小例子：找出最高分

```python
best = max(data_loaded["students"], key=lambda s: s["score"])
best
```

## 7. JSON 更新與覆寫

```python
# 更新 Amy 的成績
for s in data_loaded["students"]:
    if s["name"] == "Amy":
        s["score"] = 90

with open("class.json", "w", encoding="utf-8") as f:
    json.dump(data_loaded, f, ensure_ascii=False, indent=2)
```
