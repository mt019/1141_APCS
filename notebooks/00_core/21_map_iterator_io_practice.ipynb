這份課件目標：把 `map()`、iterator（迭代器）、以及 APCS 常見輸入格式一次打通。

你會反覆看到這句：

```python
map(int, input().split())
```

它同時牽涉到三件事：
1. `input()`：讀一行字串
2. `.split()`：把字串切成 token（字串）
3. `map(int, ...)`：把每個 token「映射」成整數（回傳 iterator）

> 注意：本課件以 ZeroJudge 直接可用的 `input()` 寫法為主。每個 code cell 上方會附「範例輸入」註解。


## 0) 先記一句話

- `split()`：字串 ➜ 很多字串
- `map(int, ...)`：很多字串 ➜ 很多整數（但它是 **iterator**）

所以：
- 你可以直接解包：`a, b = map(int, input().split())`
- 或轉成 list：`nums = list(map(int, input().split()))`


## 1) 輸入專項（最常用 3 種）

### 1-1 同一行多個數字


````python
# 範例輸入：
# 5 7

a, b = map(int, input().split())
print('a =', a)
print('b =', b)
print('a+b =', a + b)
````

### 1-2 固定筆數（先給 n，後面 n 行）

常見：n 筆資料、n 次操作、n 行座標。


````python
# 範例輸入：
# 3
# 10 1
# 20 2
# 30 3

n = int(input())
ans = []
for _ in range(n):
    x, y = map(int, input().split())
    ans.append(x + y)

print(ans)
````

### 1-3 多筆到 EOF（最常見陷阱）

ZeroJudge 很多題目沒告訴你幾筆：要讀到 EOF。

EOF（End Of File）代表「沒有更多輸入」：在測資檔到末尾時，`input()` 會丟出 `EOFError`。
在線上評測常見「多筆到 EOF」的題型，所以用 `try/except` 讀到錯誤為止。

標準模板：

```python
while True:
    try:
        line = input()
    except EOFError:
        break
    data = line.split()
    if not data:
        continue
    nums = list(map(int, data))
```


````python
# 範例輸入（EOF 結束）：
# 1 2 3
#
# 10 20
# 7

rows = []
while True:
    try:
        line = input()
        print(line)
    except EOFError:
        break

    data = line.split()
    print(data)
    if not data:
        continue

    nums = list(map(int, data))
    print(nums)
    rows.append(nums)

print(rows)
````

## 2) iterator 概念（map / zip 都是 iterator）

### 2-1 你以為你拿到的是 list，但其實不是

很多人會寫：

```python
nums = map(int, input().split())
print(nums)
```

結果印出來是一個物件，不是數字。

重點：`map(...)` 回傳的是 iterator，你要：
- `list(nums)` 看內容
- 或用 `for` 走訪


````python
# 範例輸入：
# 1 2 3

nums = map(int, input().split())
print('nums =', nums)
print('list(nums) =', list(nums))
````

### 2-2 iterator 只能走一次（「印兩次」反例）

這是最常見的觀念斷點：

- 你 `list(nums)` 一次後，iterator 就被消耗完了
- 第二次再 `list(nums)` 就會變空

這個現象同樣會出現在 `zip(...)`。


````python
# 範例輸入：
# 4 5 6

nums = map(int, input().split())
print(list(nums))
print(list(nums))  # 第二次會是空的
````

### 2-3 什麼時候該用 list(...)？什麼時候不需要？

- 你只需要「用一次」：不轉 list 也可以（省記憶體）
- 你需要多次使用 / 要 index / 要 len：就轉 list

下面示範：只做加總時，不必存 list。


````python
# 範例輸入：
# 10 20 30 40

s = sum(map(int, input().split()))
print(s)
````

## 3) 小型計算題（把 map 與 zip 串起來）

### 3-1 兩個序列的點積（dot product）

題意：兩行，各有 n 個數字，輸出 `a1*b1 + a2*b2 + ...`。

示範重點：
- 用 `map(int, ...)` 解析輸入
- 用 `zip` 讓兩個 list 並行走訪


````python
# 範例輸入：
# 1 2 3
# 4 5 6

A = list(map(int, input().split()))
B = list(map(int, input().split()))

dot = sum(a * b for a, b in zip(A, B))
print(dot)
````

### 3-2 同一行讀很多數：最大值、最小值、平均

這類題很常見：給一行數字，問 max/min/sum。


````python
# 範例輸入：
# 8 3 10 5

nums = list(map(int, input().split()))
print('max =', max(nums))
print('min =', min(nums))
print('sum =', sum(nums))
print('avg =', sum(nums) / len(nums))
````

## 4) 資料清理：map 不只用來轉 int

### 4-1 去掉頭尾空白：strip

`strip()` 很常用在：
- 讀字串題
- 讀到可能有空白的行


````python
# 範例輸入：
#   apple  
# banana
#   cat

lines = []
while True:
    try:
        line = input()
    except EOFError:
        break
    lines.append(line)

cleaned = list(map(str.strip, lines))
print(cleaned)
````

### 4-2 token 清理後再轉 int（先 strip 再 int）

如果你遇到怪格式（例如 token 旁邊有空白），可以分兩步：

1) `map(str.strip, tokens)`
2) `map(int, ...)`

（大多數正常輸入不用這麼做，但這個例子用來建立「map 是可串接的」概念。）


````python
# 範例輸入（EOF 結束）：
#  1   2
# 3    4

pairs = []
while True:
    try:
        line = input()
    except EOFError:
        break

    tokens = line.split()
    tokens = map(str.strip, tokens)
    nums = list(map(int, tokens))
    if len(nums) == 2:
        pairs.append(tuple(nums))

print(pairs)
````

## 5) 改錯題（Fix-it）— 你要看得出哪裡錯

### 5-1 錯誤 1：把 map 當成 list 重複用

錯誤程式：

```python
nums = map(int, input().split())
print(sum(nums))
print(max(nums))  # 這行會壞掉：因為 nums 被 sum 消耗完了
```

請修正：
- 方法 A：先 `list(...)`
- 方法 B：重建 iterator（再呼叫一次 map）


````python
# 範例輸入：
# 1 2 3

nums = list(map(int, input().split()))
print(sum(nums))
print(max(nums))
````

### 5-2 錯誤 2：行內數量不固定卻用 a,b 解包

錯誤程式：

```python
a, b = map(int, input().split())
```

如果輸入可能是 1 個數、或 3 個數，就會炸。

修正方向：
- 先用 `nums = list(map(int, input().split()))`
- 再依題意處理


````python
# 範例輸入：
# 10 20 30

nums = list(map(int, input().split()))
print('count =', len(nums))
print(nums)
````
