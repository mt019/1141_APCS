## 題目說明
- 題目來源：[LeetCode 912. Sort an Array](https://leetcode.com/problems/sort-an-array/description/)
- 輸入：任意整數陣列 `nums`
- 輸出：`nums` 升冪排序後的新陣列
- 評測：需在 LeetCode 的時間/記憶體限制內通過（Python 方案建議 `O(n log n)` 或最佳化 `O(n^2)` 演算法）


```python
from typing import Callable, Dict, List

SampleCase = Dict[str, List[int]]

def clone(data: List[int]) -> List[int]:
    """建立淺拷貝，避免排序函式改動原本測試資料。"""
    return list(data)

sample_cases: SampleCase = {
    'LeetCode 範例': [5, 1, 1, 2, 0, 0],
    '混合數字': [9, 4, 1, 7, 9, 3, 8, 5, 2],
    '已排序': list(range(10)),
    '逆序': list(range(10, 0, -1)),
}

def verify_sort(name: str, sorter: Callable[[List[int]], List[int]]) -> None:
    """對所有測試資料進行驗證，若失敗會觸發 AssertionError。"""
    for case_name, data in sample_cases.items():
        result = sorter(clone(data))
        assert result == sorted(data), f"{name} failed on {case_name}: {data}"
    print(f"[OK] {name} 通過 {len(sample_cases)} 組測試資料")
```

### 氣泡排序（Bubble Sort）
- 核心概念：相鄰元素兩兩比較，較大的值向右交換，逐回合將最大值「浮」到尾端。
- **時間複雜度**
  - 最佳：`O(n)`（輸入已排序，且檢測到無交換即可提前結束）
  - 平均：`O(n^2)`
  - 最差：`O(n^2)`（完全逆序）
- **空間複雜度**：`O(1)`（就地交換）
- **穩定性**：穩定（相等元素相對位置不變）
- 建議用於：資料量小或幾乎排好時的教學示範


```python
def bubble_sort(nums: List[int]) -> List[int]:
    arr = nums[:]
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

verify_sort('Bubble Sort', bubble_sort)
print('排序結果（LeetCode 範例）:', bubble_sort(sample_cases['LeetCode 範例']))
```

### 雞尾酒排序（Cocktail Shaker Sort）
- 核心概念：氣泡排序的雙向版本，先從左到右將最大值推到尾端，再從右到左將最小值推回前端。
- **時間複雜度**
  - 最佳：`O(n)`（資料已排序）
  - 平均：`O(n^2)`
  - 最差：`O(n^2)`
- **空間複雜度**：`O(1)`
- **穩定性**：穩定
- 建議用於：資料幾乎排好且小範圍逆序，雙向掃描可加速收斂


```python
def cocktail_sort(nums: List[int]) -> List[int]:
    arr = nums[:]
    start, end = 0, len(arr) - 1
    swapped = True
    while swapped:
        swapped = False
        for i in range(start, end):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swapped = True
        end -= 1
        if not swapped:
            break
        swapped = False
        for i in range(end, start, -1):
            if arr[i - 1] > arr[i]:
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                swapped = True
        start += 1
    return arr

verify_sort('Cocktail Shaker Sort', cocktail_sort)
print('排序結果（LeetCode 範例）:', cocktail_sort(sample_cases['LeetCode 範例']))
```

### 直接插入排序（Insertion Sort）
- 核心概念：維持左半邊為已排序區，將新元素插入對應位置。
- **時間複雜度**
  - 最佳：`O(n)`（資料已排序，只需一次比較）
  - 平均：`O(n^2)`
  - 最差：`O(n^2)`（逆序，每次需移動整段）
- **空間複雜度**：`O(1)`
- **穩定性**：穩定
- 建議用於：資料量小、幾乎排序好的情境，或作為複合排序（例如 Timsort）的子程序


```python
def insertion_sort(nums: List[int]) -> List[int]:
    arr = nums[:]
    for i in range(1, len(arr)):
        current = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > current:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = current
    return arr

verify_sort('Insertion Sort', insertion_sort)
print('排序結果（LeetCode 範例）:', insertion_sort(sample_cases['LeetCode 範例']))
```

### 堆排序（Heap Sort）
- 核心概念：將資料建成最小（或最大）堆，反覆彈出根節點達成排序。
- **時間複雜度**：最佳、平均、最差皆為 `O(n log n)`（建堆 `O(n)` + `n` 次 `log n` 的彈出）
- **空間複雜度**：`O(1)`（若使用原地堆化）或 `O(n)`（若使用 Python `heapq` 的最小堆副本）
- **穩定性**：不穩定
- 建議用於：需要保證 `O(n log n)` 的排序需求；在 Python 可用 `heapq` 快速示範優先權佇列


```python
import heapq

def heap_sort(nums: List[int]) -> List[int]:
    heap = list(nums)
    heapq.heapify(heap)
    return [heapq.heappop(heap) for _ in range(len(heap))]

verify_sort('Heap Sort (heapq)', heap_sort)
print('排序結果（LeetCode 範例）:', heap_sort(sample_cases['LeetCode 範例']))
```
