## 1. 什麼時候需要它？
- list/tuple 很快，但 `x[3]` 容易看不懂
- 需要 **可讀性** 與 **欄位名稱** 時，適合用 NamedTuple / dataclass

## 4. NamedTuple vs dataclass：差異一覽
**核心差異**
- **可變性**：NamedTuple 不可變（欄位不能改）；dataclass 預設可變
- **可讀性**：兩者都用 `obj.field` 讀取
- **使用場景**：
  - **資料固定**（像題目輸入、座標、考試成績）→ NamedTuple
  - **資料會更新**（像計分板、狀態）→ dataclass

**簡單判斷**
- 需要改值？→ dataclass
- 不需要改值？→ NamedTuple

## 2. NamedTuple：不可變、快速、可讀

```python
from typing import NamedTuple

class Student(NamedTuple):
    name: str
    score: int

s = Student('Mathias', 87)
print(s.name, s.score)
```

### 先解釋：什麼是曼哈頓距離
在方格地圖中，每次只能上下左右移動時，兩點間的距離就是：

`|x1 - x2| + |y1 - y2|`

它不算斜線（對角線）距離，只算「走格子」的總步數。


### 小例子：用 NamedTuple 表示座標

```python
class Point(NamedTuple):
    x: int
    y: int

p = Point(3, 4)
q = Point(10, 2)

# 計算曼哈頓距離
manhattan = abs(p.x - q.x) + abs(p.y - q.y)
print(manhattan)
```

## 3. dataclass：可變、可設預設值、可更新

```python
from dataclasses import dataclass

@dataclass
class StudentDC:
    name: str
    score: int

s = StudentDC('Lina', 92)
print(s.name, s.score)

s.score += 3
print(s)
```

### 小例子：用 dataclass 做計分板

```python
@dataclass
class ScoreBoard:
    name: str
    win: int = 0
    lose: int = 0

board = ScoreBoard('TeamA')
board.win += 1
board.lose += 2
print(board)
```

## 4-1. 何時用 NamedTuple / dataclass（更多對比例子）
以下每組都是「同一情境」的兩種做法，對比差異。


```python
# 情境 1：考卷成績（不變） -> NamedTuple
class Exam(NamedTuple):
    name: str
    score: int

e = Exam("Midterm", 85)
print(e)
# e.score = 90  # 會報錯，因為不可變
```

```python
# 情境 1：考卷成績（要改分數） -> dataclass
@dataclass
class ExamDC:
    name: str
    score: int

e = ExamDC("Midterm", 85)
e.score = 90
print(e)
```

```python
# 情境 2：座位表（固定） -> NamedTuple
class Seat(NamedTuple):
    row: int
    col: int
    name: str

s = Seat(2, 5, "Amy")
print(s)
```

```python
# 情境 2：座位表（會換座位） -> dataclass
@dataclass
class SeatDC:
    row: int
    col: int
    name: str

s = SeatDC(2, 5, "Amy")
s.row, s.col = 3, 1
print(s)
```

```python
# 情境 3：公交站點座標（固定） -> NamedTuple
class Stop(NamedTuple):
    name: str
    x: int
    y: int

stop = Stop("A1", 10, 7)
print(stop)
```

```python
# 情境 3：遊戲角色座標（會移動） -> dataclass
@dataclass
class Player:
    name: str
    x: int
    y: int

p = Player("Hero", 10, 7)
p.x += 1
p.y -= 2
print(p)
```

```python
# 情境 4：題目輸入資料（固定） -> NamedTuple
class Query(NamedTuple):
    a: int
    b: int

q = Query(3, -1)
print(q)
```

```python
# 情境 4：處理狀態（會更新） -> dataclass
@dataclass
class State:
    step: int
    value: int

st = State(0, 10)
st.step += 1
st.value *= 2
print(st)
```

## 6. 高中生熟悉的例子
用日常情境來看 NamedTuple / dataclass 的好處。


```python
# 例子 1：段考成績（不會改 -> NamedTuple）
class ExamScore(NamedTuple):
    subject: str
    score: int

scores = [ExamScore("Math", 88), ExamScore("English", 92)]
avg = sum(s.score for s in scores) / len(scores)
avg
```

```python
# 例子 2：課表（資料固定 -> NamedTuple）
class ClassSlot(NamedTuple):
    day: str
    period: int
    subject: str

slot = ClassSlot("Mon", 3, "Physics")
slot
```

```python
# 例子 3：社團出席（會更新 -> dataclass）
@dataclass
class Attendance:
    name: str
    present: int = 0
    absent: int = 0

a = Attendance("Leo")
a.present += 1
a.absent += 2
a
```

```python
# 例子 4：飲料訂單（會修改數量 -> dataclass）
@dataclass
class DrinkOrder:
    name: str
    size: str
    qty: int = 1

order = DrinkOrder("BlackTea", "M")
order.qty += 2
order
```

## 5. 與本題骰子題的連結
- 用 list 會出現 `dice[0]`, `dice[1]` 等難讀索引
- NamedTuple / dataclass 可以用 `dice.top`、`dice.front` 提升可讀性

```python
# NamedTuple 版本（只示範資料結構）
class Dice(NamedTuple):
    top: int
    bottom: int
    front: int
    back: int
    right: int
    left: int

base = Dice(1, 6, 4, 3, 2, 5)
print(base.top, base.front, base.right)
```
