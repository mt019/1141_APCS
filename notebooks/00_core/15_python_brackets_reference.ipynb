## 1. `()` 圓括號：函式呼叫、群組運算與 tuple

常見用途：

- **函式呼叫**：`print("Hello")`、`len(data)`。
- **運算優先順序**：`(a + b) * c`。
- **tuple 字面值**：`point = (25.0, 121.5)`；若想建立單元素 tuple，需寫成 `(value,)`。
- **生成器表達式 / 內建函式參數**：`sum(x for x in numbers)`。

下方程式展示幾種常見形式，並標註圓括號的功能。

```python
# 函式呼叫：() 包住引數
message = "Hello"
print("大寫：", message.upper())  # upper() 呼叫字串方法

# 運算優先順序：() 改變計算順序
a, b, c = 2, 3, 4
print("(a + b) * c =", (a + b) * c)
print("a + (b * c) =", a + (b * c))

# tuple 字面值：() 建立不可變序列
coords = (25.0478, 121.5319)
lat, lng = coords  # tuple unpacking
print("Lat, Lng =", lat, lng)

# 單元素 tuple 需加逗號
single = (42,)
print("single 的型別：", type(single))

# 生成器表達式：() 可搭配 sum / any / all
numbers = [1, 2, 3, 4, 5]
total = sum(x * x for x in numbers if x % 2 == 1)
print("奇數平方和 =", total)
```

```python
# sum([1, 2, 3, 4, 5])  # 使用 list
# sum((1, 4, 9, 16, 25))  # 使用 tuple
sum(x * x for x in (2,4))  # 使用生成器表達式
```

## 2. `[]` 方括號：list、索引、切片與 list comprehension

常見用途：

- **list 字面值**：`days = ["Mon", "Tue", "Wed"]`。
- **索引與切片**：`days[0]`、`days[1:3]`。
- **可變結構：list 支援新增、刪除、指派**。
- **list comprehension**：`[x * x for x in range(5)]`。
- **在字典或串列內嵌結構**：`matrix[0][2]`。

下方程式示範方括號的各種使用情境。

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
print('第一天：', days[0])
print('周中：', days[2:4])  # 切片 (不包含 index 4)

# list 可修改：
days[0] = 'Monday'
days.append('Sat')
print('更新後的 days：', days)

# list comprehension：
lengths = [len(name) for name in days if name.startswith(('S', 'M'))]
print('特定開頭的名字長度：', lengths)

# 巢狀 nested list / 索引：
schedule = [
    ['math', 'english', 'pe'],
    ['chemistry', 'art', 'music'],
]
print('第二天第三節課：', schedule[1][2])
```

## 3. `{}` 花括號：dict 與 set

常見用途：

- **dict 字面值**：`user = {'name': 'Iris', 'tries': 2}`，使用 `key: value` 配對。
- **set 字面值**：`tags = {'python', 'boolean'}`，集合元素不重複、無順序。
- **comprehension**：`{key: value for ...}`、`{expr for ...}`。
- **空花括號預設為 dict**；若要建立空集合需寫 `set()`。

下方程式展示 dict 與 set 的典型操作，並比較它們的特性。

```python
profile = {'name': 'Iris', 'tries': 2}
print('名稱 =>', profile['name'])
profile['tries'] += 1
profile['last_login'] = '2025-10-22'
print('更新後 profile =>', profile)

tags = {'python', 'loop', 'boolean'}
tags.add('login')
tags.add('python')  # 重複值不會改變集合
print('tags =>', tags)

# dict comprehension：統計字串長度
words = ['break', 'continue', 'pass']
length_map = {word: len(word) for word in words}
print('長度對照 =>', length_map)

# set comprehension：挑出長度 > 4
long_words = {word for word in words if len(word) > 4}
print('長字集合 =>', long_words)
```

### `{}` 在字串格式化中的角色

`{}` 除了建立 dict / set，也常用在字串格式化中，作為佔位符（placeholder）。主要有三種常見形式：

1. `str.format()`：把值依序或依名稱填入 `{}`。
2. f-string：在 `{}` 中直接寫表達式。
3. 與舊版 `%` 格式不同，`{}` 可以指定格式規則，例如 `{:>10}` 代表右對齊。

下方程式比較 `str.format()` 與 f-string 的使用方式，順便示範轉義大括號的方法。

```python
name = 'Iris'
tries = 3
success_rate = 0.875

# 1. str.format()：位置或名稱對應
template = 'User {0} has tried {1} times. Success rate = {rate:.1%}'
print(template.format(name, tries, rate=success_rate))

print('User {1} has tried {0} times. Success rate = {rate:.2%}'.format(tries, name, rate=success_rate))

# 2. f-string：直接在 {} 中寫表達式
print(f'User {name} has tried {tries} times. Success rate = {success_rate:.4}')

# 3. 轉義大括號：使用 {{ 和 }} 顯示字面上的 {}
print('Display literal braces: {{example}} -> {}'.format('示範'))
print(f'Display literal braces in f-string: {{example}}')

# 4. 指定格式規則
print('右對齊寬度 10：{:>10}'.format(name))
print(f'左對齊寬度 10：{name:<10}!')
```

## 4. 綜合比較表與練習題

| 括號 | 典型資料結構 / 語意 | 是否可變 | 是否保順序 | 是否允許重複 | 常見操作 |
| --- | --- | --- | --- | --- | --- |
| `()` | tuple、函式呼叫、運算 grouping | 否（tuple） | 是（tuple） | 是 | unpack、函式呼叫、生成器 |
| `[]` | list、索引、切片、list comprehension | 是 | 是 | 是 | append、index、slice |
| `{}` | dict（key: value）、set、comprehension | dict：是；set：是 | dict：維持插入順序；set：無順序 | dict：鍵不可重複；set：元素不重複 | keys、values、add、update |

### 撰寫 `summarize_attempts` 之前的語法鋪墊
1. **迴圈拆封 tuple**：`for name, success in log` 會把每筆 `(name, success)` 自動拆成兩個變數，等同於 `pair = (name, success); name = pair[0]`。
2. **dict 累計次數**：`stats[name] = stats.get(name, 0) + 1` 利用 `dict.get` 取得既有值，若鍵不存在則回傳預設 0，方便做加總。
3. **set 儲存唯一元素**：`success_set.add(name)` 只在成功時加入學生姓名，set 自動過濾重複值。常見於需要「集合」概念的題目。
4. **條件判斷**：`if success:` 表示只有當成功布林值為 `True` 時才執行區塊，不需要寫成 `if success == True`。
5. **多值回傳 (tuple)**：`return stats, success_set` 會組成一個二元素 tuple，方便同時回傳 dict 與 set，呼叫端可以用 `stats, success = summarize_attempts(log)` 一次接住。
> 建議先獨立練習每個語法片段，再回來組合成完整函式。這些技巧也會在後續的資料分析實作題中頻繁出現。


```python
sample_log = [
    ('Amy', True),
    ('Ben', False),
    ('Amy', False),
    ('Cindy', True),
]

stats = {}
success_set = set()

for name, success in sample_log:
    print('處理項目:', (name, success))
    stats[name] = stats.get(name, 0) + 1
    print('  -> stats 現在是', stats)
    if success:
        success_set.add(name)
        print('  -> success_set 加入成功者', success_set)
    else:
        print('  -> success 為 False，success_set 維持', success_set)

print('函式最終會回傳:', (stats, success_set))
```

### 還有哪些常見卻未細談的語法？
- **tuple 拆封與星號運算子**：`a, *mid, b = values` 可一次取頭尾與中間資料；`return tuple_obj` 常搭配函式多值回傳。
- **序列 slicing 變化**：`seq[start:stop:step]`、負索引、倒序 `seq[::-1]`，對 list / tuple 同樣適用。
- **list comprehension 進階技巧**：巢狀 comprehension、條件表達式 `[(x if x > 0 else 0) for x in data]`。
- **enumerate / zip**：`for idx, item in enumerate(list_obj)` 保留索引；`zip(names, scores)` 一次遍歷多個序列。
- **dict 常用方法**：`dict.get(key, default)`、`setdefault`、`items()` / `values()` 迭代；`dict` comprehension `{k: v for k, v in pairs if v > 0}`。
- **set 運算**：聯集 `|`、交集 `&`、差集 `-`、對稱差 `^`，以及 `subset` / `superset` 判斷。
- **collections 模組**：`defaultdict` 自動初始化、`Counter` 統計次數、`OrderedDict` 保留插入順序。
- **解包到函式引數**：`func(*args_list)`、`func(**kwargs_dict)`，分別對應序列與 dict 的拆封傳參。
- **不可變與複製議題**：`list.copy()` / `dict.copy()` 是淺拷貝；巢狀結構可考慮 `copy.deepcopy()`。
