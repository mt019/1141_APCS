本筆記收錄 APCS 練習常見的 Python 語法：資料型別、控制流程、函式、常見容器操作與簡短練習。


## 1. 基本資料型別與輸出 Basic Types & Printing


````python
# 整數、浮點數、字串與布林
num = 42
pi = 3.1415
name = 'Alice'
flag = True

print(num, type(num))
print(pi, type(pi))
print(name, type(name))
print(flag, type(flag))
````

- 使用 `type()` 確認型別。
- 字串可用單引號或雙引號。
- 布林值在條件判斷中最常見。


## 2. 輸入與型別轉換 Input & Casting


````python
raw = input('輸入一個整數: ')  # APCS 常見格式：先讀入、再轉型別
value = int(raw)
print('平方:', value ** 2)
````

- `input()` 讀進來的是字串，需視需求轉為 `int` 或 `float`。
- APCS 題目常搭配 `split()` 讀多個數字：`a, b = map(int, input().split())`。


## 3. 條件判斷 Conditionals


````python
score = 75
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'D'
print('等第:', grade)
````

- `if / elif / else` 結構會依序判斷。
- 比較運算可鏈結：`0 <= x <= 100`。


## 4. 迴圈 Loops


### 4.1 `for` 迴圈


````python
type(range(1,6))
print(range(1,6))
print(list(range(1,6)))
````

````python
total = 0
for i in range(1, 6):  # 1 到 5
    total += i

print('1...5 相加 =', total)
````

````python
total = 0
for i in range(2, 8, 2):  
    total += i
    print(i)

print('1+3+...+7 相加 =', total)
````

````python
total = 0
for i in range(10, 0, -1):  
    total += i
    print(i)

print('total =', total)
````

- `range(start, end, step)`，不包含 `end`。
- 列表推導式 `squares = [x*x for x in range(10)]` 常見於 APCS。


### 4.2 `while` 迴圈


````python
count = 0
while count < 5:
    print('loop', count)
    count += 1
````

- 使用 `while` 需注意終止條件避免無窮迴圈。
- `break` 可提前跳出，`continue` 跳過當前循環。


## 5. 函式 Functions


````python
def factorial(n):
    '''計算 n!，假設 n >= 0'''
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

print(factorial(5))
````

- 使用 `return` 回傳值。
- 函式可預設參數，例如 `def power(x, y=2): return x ** y`。
- 遞迴函式在 APCS 題目也常出現，需要留意終止條件。


## 6. 常見容器與操作 Containers


### 6.1 List 列表


````python
nums = [3, 1, 4, 1, 5]
nums.append(9)
nums.sort()  # 就地排序
print(nums)
print('第一個元素:', nums[0])
print('最後一個:', nums[-1])
````

- 常用方法：`append`, `pop`, `sort`, `reverse`, `count`, `index`。
- 切片 `nums[1:4]` 可取得子列表。


### 6.2 Tuple 與 Set


````python
point = (3, 4)
unique = {3, 1, 4, 1, 5}
print(point, type(point))
print(unique, type(unique))
````

- Tuple 為不可變序列，適合保存固定組合。
- Set 自動去除重覆，適合快速檢查元素存在性 (`in` 操作)。


### 6.3 Dictionary 字典


````python
scores = {'Alice': 90, 'Bob': 80}
scores['Cindy'] = 95
for name, score in scores.items():
    print(name, score)
````

- 透過 `keys()`, `values()`, `items()` 遍歷。
- APCS 題目常用字典統計次數或建立映射。


## 7. List Comprehension & Generator


````python
squares = [x*x for x in range(10)]
print('平方列表:', squares)

even_gen = (x for x in range(20) if x % 2 == 0)
print('前四個偶數:', [next(even_gen) for _ in range(4)])
````

- 列表推導式可將迴圈與條件壓縮成一行。
- 生成器節省記憶體，需透過 `next()` 取得值。


## 8. 字串處理 String Operations


````python
text = 'APCS Python'
print(text.lower())
print(text.split())
print('-'.join(['A', 'B', 'C']))
````

- 常用方法：`strip`, `split`, `join`, `replace`, `find`。
- 字串也是序列，可用 `text[0]`、`text[::-1]` 等切片。


## 9. 模組與常用庫 Modules & Libraries


````python
import math
import statistics as stats

print('sqrt(2)=', math.sqrt(2))
print('mean:', stats.mean([1,2,3,4,5]))
````

- 常見標準庫：`math`, `statistics`, `collections` (`Counter`, `deque`), `heapq`。
- 使用 `from collections import Counter` 可快速統計。


## 10. 簡單練習 Mini Exercises


````python
def is_prime(n):
    if n < 2: 
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print([x for x in range(2, 30) if is_prime(x)])
````

- 練習：嘗試輸入一串數字，輸出其中的偶數 / 奇數總和。
- 練習：寫一個函式判斷字串是否迴文。
