# Flowchart 入門筆記

這份 Notebook 用來介紹程式流程圖（flowchart）的基本符號與示例，協助學生把口頭描述或程式碼視覺化。可搭配圖卡、便利貼或白板活動使用。

## 1. 為什麼要畫流程圖？

- 把程式流程視覺化，快速看出是否有遺漏某個步驟或出口。
- 更容易討論「先做什麼、判斷什麼、再做什麼」，對於拆解題目特別有幫助。
- 在寫程式之前先畫流程圖，可以提早發現無限迴圈、條件重複等問題。

## 2. 常見流程圖符號

| 符號 | 名稱 | 功能 | 對應程式語法 |
| --- | --- | --- | --- |
| ⏺ / ⭕ | 起點 / 終點 (Terminator) | 表示流程開始或結束 | `程式開始`、`return`、`break` 後的結束 |
| ▭ | 流程 (Process) | 執行動作或賦值 | `total = 0`、`tries += 1` |
| ⪢ | 輸入 / 輸出 (I/O) | 取得輸入或顯示結果 | `input()`、`print()` |
| ◇ | 決策 (Decision) | 依條件分支 | `if` / `while` 條件，True / False 兩條路 |
| ⇄ | 箭頭 (Flow line) | 連結節點，指向下一個步驟 | `while` 迴圈回頭、`if` 分支指向不同路徑 |

### 小提醒
- **每個流程圖至少要有開始與結束**，否則可能出現無限循環。
- 決策（◇）一定要標出兩條路徑，通常是 `True` / `False`。
- 當多個步驟重複時，可用迴圈箭頭回到決策或流程節點。

## 3. 範例：密碼最多輸入三次

以下用簡化文字描繪流程。課堂上建議使用圖卡或繪圖工具，把每個節點貼在白板上再拉線連結。

```text
[開始]
  |
  v
tries = 0
  |
  v
讀取使用者輸入
  |
  v
密碼正確?
 /    \
是      否
 |        |
 v        v
[登入成功]  tries += 1
             |
             v
       tries < max?
         /    \
       是      否
       |        |
       v        v
  讀取下一次   [登入失敗]
```

### 對照 `while` + `break` 的程式碼
```python
tries = 0
max_tries = 3
while tries < max_tries:
    user_input = input('請輸入密碼：')
    if user_input == password:
        print('登入成功')
        break
    tries += 1
else:
    print('登入失敗')
```

## 4. 從描述到流程圖，再到程式碼

1. **文字描述**：先寫出步驟，例如「重複詢問密碼，最多三次」。
2. **列出關鍵事件**：初始化、取得輸入、檢查條件、更新計數、輸出結果。
3. **安排順序與分支**：決定哪些步驟循環、哪些是一次性。
4. **畫流程圖**：使用符號代表步驟與判斷。
5. **轉成程式碼**：把每個節點對應到 `while`、`if`、`print()` 等語句。
6. **檢查對照**：沿著流程圖走一次，確認程式會走到的每條路徑都在程式碼中出現。

## 5. 課堂活動建議

- **圖卡拼拼樂**：準備印有符號的卡片與程式片段，讓學生把對應的步驟放在正確的符號上。
