## 1. 日常語意 → 布林判斷

先把句子拆成兩個布林變數（A、B）。True 表示事件發生。接著列出 4 種組合並計算：
- `A and B`：兩者都為 True 時才 True。
- `A or B`：至少有一個 True 就 True。
- `A xor B`：只有一個 True（不相同）時才 True。

完成真值表後，再整理出符合語句的列，寫成 Python 條件。

### 範例：下雨我帶傘（A=`is_raining`, B=`bring_umbrella`）
| A | B | A and B | A or B | A xor B | 你的判斷 |
|---|---|----------|---------|----------|-----------|
| False | False | False | False | False |  不下雨、不帶傘 |
| False | True  | False | True  | True  |  不下雨但帶傘 |
| True  | False | False | True  | True  |  下雨沒帶傘 |
| True  | True  | True  | True  | False |  下雨且帶傘 |

- Python 可能寫法：
  - `is_raining and bring_umbrella`
  - 或 `not is_raining or bring_umbrella`

| A | B | A and B | A or B | A xor B | 你的判斷 |

| False | True  | | | | |
| True  | False | | | | |
| True  | True  | | | | |
- Python 寫法：__________


````python
# 生成情境真值表的範例程式
from itertools import product

def truth_table_for(statement_fn, headers):
    name_cols = ' '.join('{:>12}'.format(h) for h in headers)
    print('{} | 結果'.format(name_cols))
    for combo in product([False, True], repeat=len(headers)):
        result = statement_fn(*combo)
        values = ' '.join('{:>12}'.format(str(v)) for v in combo)
        print('{} | {}'.format(values, result))

def went_swim_after_icecream(is_raining, ate_icecream, go_swim):
    return ate_icecream and go_swim

truth_table_for(went_swim_after_icecream, ('is_raining', 'ate_icecream', 'go_swim'))
````

## 3.5 `while` 迴圈複習

`while` 會重複執行程式區塊，直到條件為 False。請記得：
1. 迴圈之前設定好初始值。
2. 迴圈內要更新條件相關變數。
3. 條件終究要變成 False，否則會無限迴圈。

### 範例：從 0 數到 4
```python
count = 0
while count < 5:
    print('第', count, '次')
    count += 1
```

| 回合 | 進入前 `count` | 判斷 `count < 5` | 執行內容 | 更新後 `count` |
|------|-----------------|------------------|-----------|----------------|
| 1 | 0 | True | 印出 `第 0 次` | 1 |
| 2 | 1 | True | 印出 `第 1 次` | 2 |
| 3 | 2 | True | 印出 `第 2 次` | 3 |
| 4 | 3 | True | 印出 `第 3 次` | 4 |
| 5 | 4 | True | 印出 `第 4 次` | 5 |
| 6 | 5 | False | 不進入迴圈 | 5 |

- `break`：在特定條件下立刻離開迴圈。
- `continue`：跳過本輪剩餘程式，直接回到條件判斷。

> 練習：把自己的 `while` 迴圈也用表格記錄下來。


````python
# 試著自行記錄 while 迴圈每一輪的狀態
count = 0
history = []
while count < 5:
    history.append({'before': count, 'condition': count < 5})
    print('第', count, '次')
    count += 1
    history[-1]['after'] = count

history
````

### 階段 2：加入輸入與比較
- 取得第一次 `user_input`。
- 當 `user_input != password` 且尚未用完次數時，繼續詢問。


````python
password = '1234'
max_tries = 5
tries = 0
user_input = input('請輸入密碼：')

while user_input != password and tries < max_tries:
    tries += 1
    user_input = input('密碼錯誤，請再試一次：')

print('迴圈結束，tries =', tries)
````

### 階段 3：剩餘次數提示 + 結束訊息
- 在迴圈內計算 `remaining = max_tries - tries`。
- 迴圈結束後，用 `if/else` 判斷登入成功或鎖定。


````python
password = '1234'
max_tries = 5
tries = 0
user_input = input('請輸入密碼：')

while user_input != password and tries < max_tries - 1:
    tries += 1
    remaining = max_tries - tries
    print('密碼錯誤，剩下', remaining, '次機會')
    user_input = input('請再試一次：')

if user_input == password:
    print('登入成功，歡迎光臨！')
else:
    print('錯誤次數過多，帳號已鎖定。')
````
