## 1. Fact vs Statement：布林變數的語意基礎

在課堂中我們發現，有時候會把布林變數當成抽象名詞。為了能把生活語句轉成布林判斷，需要先分辨：

- **Fact**：已知為真或為假的具體事實，通常是客觀觀察，例如「現在正在下雨」。
- **Statement**：能夠被判真或假的句子，可以回答「是或不是」。程式裡的布林變數通常對應 Statement。

### 課堂示例：下雨與帶傘

| 代碼 | Statement（可判真假的句子） | 可能的 Fact |  判斷結果？|
| --- | --- | --- | --- |
| A | 今天正在下雨 | 親眼看到窗外下雨 |  |
| B | 我有帶傘出門 | 看到包包裡有傘 |  |


`A and B` 是否成立？
- 只有在 A 與 B 都為 True 時才為 True

完成真值表時，可以請學生逐列描述：

- `A = False, B = False`：今天天氣晴朗，而且沒帶傘 → 「下雨且帶傘」為 False。
- `A = True, B = False`：有下雨但沒帶傘 → 不符合句子。
- `A = False, B = True`：沒下雨但帶傘 → 仍然不符合句子。
- `A = True, B = True`：下雨且帶傘 → 句子成立。

### 快速檢核活動建議

1. 先讓學生用中文寫一句「可以回答是或不是」的句子，再請他們替句子命名布林變數（Statement）。
2. 請學生把同一句話拆成 Fact 與 Statement：
   - Fact：實際上觀察到的狀態。
   - Statement：我們在程式中想判斷的一句話。
3. 讓學生將同一 Statement 套用到不同 Fact（例如換成「吃冰淇淋」「去游泳」）建立第二份真值表，加深遷移。

### 課後延伸

- 請學生各自寫出 3 組 Statement，並為每組列出完整真值表。
- 要求學生說明每個欄位對應的生活語意，確保不是只在填 True/False。

```python
# 示範：用程式描述 A/B 真值語意
scenarios = [
    {"A": False, "B": False},
    {"A": True, "B": False},
    {"A": False, "B": True},
    {"A": True, "B": True},
]

def describe_scenario(a_is_true, b_is_true):
    rain_fact = "正在下雨" if a_is_true else "沒有下雨"
    umbrella_fact = "有帶傘" if b_is_true else "沒有帶傘"
    statement = a_is_true and b_is_true
    return rain_fact, umbrella_fact, statement

for item in scenarios:
    rain, umbrella, result = describe_scenario(item["A"], item["B"])
    print(f"A={item['A']:5} | B={item['B']:5} ｜ {rain}、{umbrella} ｜ A and B → {result}")
```

## 2. 迴圈控制關鍵字與模式梳理

課堂上學生主動使用 `break` 來提前離開 `while`。為了讓下週的教學更完整，先整理常見的迴圈控制方法：

| 關鍵字 / 模式 | 用法摘要 | 典型用途 | 常見陷阱 |
| --- | --- | --- | --- |
| `break` | 立即跳出最近一層迴圈 | 找到目標後停止搜尋；提前結束登入流程 | 忘記同時處理「之後要輸出的訊息」，導致 `for-else` 不會執行 |
| `continue` | 略過本輪剩餘敘述，直接進入下一輪 | 略過不合法輸入、跳過非目標資料 | 放在計數器更新之前會造成無限迴圈或錯誤統計 |
| `pass` | 什麼也不做，佔位 | 預留程式骨架、等候之後補上邏輯 | 以為會自動跳出或繼續，事實上只是空敘述 |
| `while ... else:` / `for ... else:` | 只有沒有 `break` 時才會執行 `else` 區塊 | 搜尋全部項目仍沒找到時，提供「失敗」訊息 | 忘了 `break` 也會阻止 `else` 執行 |

### 教學建議

- 每個關鍵字至少搭配一個正向範例與一個錯誤示例（例如在 `continue` 之前少了 `i += 1`）。
- 讓學生填寫「這一輪前後的變數值」，對照 `break` / `continue` 是否改變流程。
- 和計數器練習連結：請學生解釋 `count += 1` 擺在 `continue` 前後的差異。

```python
# `break` / `continue` / `for-else` 範例
numbers = [5, 3, -1, 2, 0, 4]
print("使用 break：提前停止遇到第一個負數")
for n in numbers:
    if n < 0:
        print(f"遇到 {n} → break")
        break
    print(f"處理 {n}")
else:
    print("for-else 只有在沒有 break 時才會執行")

print("\n使用 continue：略過負數，但迴圈仍持續")
for n in numbers:
    if n < 0:
        print(f"跳過 {n}")
        continue
    print(f"處理 {n}")

print("\n使用 pass：暫時保留結構")
for n in numbers:
    if n == 0:
        pass  # 之後再決定 0 要怎麼處理
    else:
        print(f"其他數字 {n}")
```

## 3. 密碼錯誤次數限制：多種寫法與比較

本週學生嘗試了不同的 `while` 寫法。以下整理常見結構、優缺點與測試方式：

| 版本 | 核心概念 | 優點 | 注意事項 |
| --- | --- | --- | --- |
| 3A. 前置輸入 + `while` 僅管控次數 | 先在迴圈前讀一次輸入，`while` 判斷「輸入錯誤」且「次數未達上限」 | 結構直觀 | 若沒把第一次輸入算進次數，會出現多給一次機會的 off-by-one |
| 3B. `while` 同時檢查兩個條件 | `while user_input != password and tries < max_tries:` | 不需要 `break`，邏輯集中 | 必須正確更新計數器，以免條件永遠 True |
| 3C. `while tries < max_tries` + `break` | 迴圈只關心次數，密碼判斷放在迴圈內，成功立刻 `break` | 最貼近「類似 for 迴圈」的思路 | 記得加上 `else` 或迴圈後的失敗訊息 |
| 3D. `for` 迴圈版本 | `for attempt in range(1, max_tries + 1)` | 自帶次數控制，程式碼短 | 需注意 `for-else` 與 `break` 的配合 |

### 模擬輸入輔助函式

以下提供一個 `make_mock_input`，用來在 Notebook 中模擬多次輸入，方便檢驗各版本在不同情境的行為。

```python
def make_mock_input(answers):
    """建立模擬 input() 的函式，依序取用 answers。"""
    answers = list(answers)
    def mock_input(prompt):
        if not answers:
            raise RuntimeError("mock_input 沒有更多資料可用")
        value = answers.pop(0)
        print(f"{prompt}{value}")
        return value
    return mock_input
```

```python
# 3A. 前置輸入版本（含常見 off-by-one）
def login_version_pre_input(mock_input, password="apcs", max_tries=3):
    tries = 0
    user_input = mock_input("第一次輸入：")
    while user_input != password and tries != max_tries:
        tries += 1
        remaining = max_tries - tries
        print(f"第 {tries} 次錯誤，剩餘 {remaining} 次")
        user_input = mock_input("請重新輸入：")
    if tries == max_tries:
        print("結果：登入失敗（注意：這版會讓使用者輸入第 4 次才跳出）")
    else:
        print("結果：登入成功")

print("--- 3A 測試：全部輸錯 ---")
login_version_pre_input(
    make_mock_input(["1111", "0000", "2222", "3333"]),
    password="apcs",
    max_tries=3,
)

print("\n--- 3A 測試：第四次才答對 ---")
login_version_pre_input(
    make_mock_input(["1111", "0000", "apcs"]),
    password="apcs",
    max_tries=2,
)
```

```python
# 3B. `while` 同時判斷密碼與次數（無 break）
def login_version_double_condition(mock_input, password="apcs", max_tries=3):
    user_input = mock_input("請輸入密碼：")
    tries = 1  # 已經使用第一次嘗試
    while user_input != password and tries < max_tries:
        remaining = max_tries - tries
        print(f"錯誤，剩餘 {remaining} 次")
        user_input = mock_input("請重新輸入：")
        tries += 1
    if user_input == password:
        print("結果：登入成功")
    else:
        print("結果：登入失敗（已用完所有次數）")

print("--- 3B 測試：第二次答對 ---")
login_version_double_condition(
    make_mock_input(["1111", "apcs"]),
    password="apcs",
    max_tries=3,
)

print("\n--- 3B 測試：全部輸錯 ---")
login_version_double_condition(
    make_mock_input(["aaaa", "bbbb", "cccc"]),
    password="apcs",
    max_tries=3,
)
```

```python
# 3C. `while tries < max_tries` + `break`
def login_version_with_break(mock_input, password="apcs", max_tries=3):
    tries = 0
    while tries < max_tries:
        user_input = mock_input(f"第 {tries + 1} 次輸入：")
        if user_input == password:
            print("結果：登入成功")
            break
        tries += 1
        remaining = max_tries - tries
        print(f"錯誤，剩餘 {remaining} 次")
    else:
        print("結果：登入失敗（用完所有次數）")

print("--- 3C 測試：第三次成功 ---")
login_version_with_break(
    make_mock_input(["1111", "2222", "apcs"]),
    password="apcs",
    max_tries=3,
)

print("\n--- 3C 測試：全部輸錯 ---")
login_version_with_break(
    make_mock_input(["aaaa", "bbbb", "cccc"]),
    password="apcs",
    max_tries=3,
)
```

```python
# 3D. `for` 迴圈版本
def login_version_for_loop(mock_input, password="apcs", max_tries=3):
    for attempt in range(1, max_tries + 1):
        user_input = mock_input(f"第 {attempt} 次輸入：")
        if user_input == password:
            print("結果：登入成功")
            break
        remaining = max_tries - attempt
        if remaining:
            print(f"錯誤，還有 {remaining} 次")
        else:
            print("錯誤，沒有剩餘次數")
    else:
        print("結果：登入失敗（for-else 發生在沒有 break 時）")

print("--- 3D 測試：第一次就成功 ---")
login_version_for_loop(
    make_mock_input(["apcs"]),
    password="apcs",
    max_tries=3,
)

print("\n--- 3D 測試：全部輸錯 ---")
login_version_for_loop(
    make_mock_input(["1111", "2222", "3333"]),
    password="apcs",
    max_tries=3,
)
```

### Debug checklist

1. **計數器是否在每次錯誤後更新？** 如果用了 `continue`，要確認更新仍然會執行。
2. **條件是否會在某個時刻變 False？** 檢查 `while` 條件中每個變數是否在迴圈內有被改變。
3. **訊息對應的事件是否正確？** 例如 `tries == max_tries` 時要避免印出「登入成功」。
4. **成功與失敗的出口是否互斥？** 建議把成功與失敗的輸出包在互斥的 `if/else` 或 `break` + `else` 中。
5. **測試多種情境：** 一次成功、最後一次成功、全部失敗、沒有輸入資料、輸入不足等。
