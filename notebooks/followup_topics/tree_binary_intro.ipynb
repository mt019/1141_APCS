# 樹與二元樹基礎講義

> 本講義將帶你從「樹」的基本觀念出發，理解節點／邊、親子關係、高度／深度等術語，進一步專注在最常見的 **二元樹** 與其遍歷 (traversal)。內容包含圖文示意、程式表達方式與練習題。


## 0. 為什麼要學樹？
- 資料常呈現階層關係：檔案系統、家族譜、網站導覽、表達式。
- 許多演算法（搜尋、排序、壓縮、語法解析）都以樹作為核心資料結構。
- 理解樹結構是進入更進階題目的基礎（如二元搜尋樹、堆積、線段樹）。


## 1. 樹的基本術語
一棵樹 (Tree) 是 **無循環** 的連通圖。常見術語如下：

```
            (root)
              A
            /   \
           B     C
         / | \   | \          
        D  E  F  G  H   
        
```

- `A`：根節點 (root)。
- `B`, `C`：A 的子節點 (child)；A 是它們的父節點 (parent)。
- `D`, `E`, `F`, `G`, `H`：葉節點 (leaf)，沒有子節點。
- 節點的層級 (level) / 深度 (depth)：從根到該節點需走的邊數；例如 `A` 深度 0，`B` 深度 1，`H` 深度 3。
- 樹的高度 (height)：所有節點深度的最大值。
- 子樹 (subtree)：一個節點與其所有後代構成的樹。


## 2. 二元樹 (Binary Tree)
二元樹限制每個節點最多有兩個子節點：左子節點 (left child) 與右子節點 (right child)。

```
        8
      /   \      
     4     10
    / \      \  
   2   6      20
```

在二元樹中常用的術語：
- **滿二元樹** (Full)：節點要嘛有 0 個子節點，要嘛有 2 個子節點。
- **完全二元樹** (Complete)：除了最後一層外，其餘層都填滿，最後一層節點靠左。
- **完美二元樹** (Perfect)：所有層都填滿。


### 2.1 二元樹在程式中的表達方式
1. **類別節點 (Node Class)**：常見於教學與遞迴題。
2. **巢狀字典 / Tuple**：快速建立靜態範例。
3. **陣列索引**：適用於完全二元樹（堆積 Heap）。

下列示範第一種寫法，先回顧 Python `class` 語法：
```python
class 類別名稱:
    def __init__(self, 參數...):
        # 這裡放初始化程式碼
```
- `class Node:` 宣告一個名為 `Node` 的自訂資料型別，通常用來描述樹的「節點」。
- `__init__` 是物件建立時會自動呼叫的建構子；`self` 代表「剛建立的那個節點本身」。
- 參數 `value`, `left=None`, `right=None` 分別對應節點儲存的值、左子節點、右子節點；預設值 `None` 代表可以先不提供子節點。
- `self.value = value` 這類賦值動作會把資料存進節點物件裡，如此後續才能透過 `node.value` 取用。

理解以上語法後，再看下一個程式範例。

````python
class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

# 建立上圖的樹
root = Node(8,
            left=Node(4, Node(2), Node(6)),
            right=Node(10, right=Node(20)))

print(root.value, root.left.value, root.right.value)
````

## 3. 樹的遍歷 (Tree Traversal)
遍歷指的是「按照某種順序拜訪樹的每個節點」。在二元樹中特別常見四種：
1. **前序 (Preorder)**：`節點 → 左子樹 → 右子樹`
2. **中序 (Inorder)**：`左子樹 → 節點 → 右子樹`
3. **後序 (Postorder)**：`左子樹 → 右子樹 → 節點`
4. **層序 (Level-order)**：逐層掃描（通常使用 BFS）

以 `8-4-10` 的例子為例：
- 前序：`8, 4, 2, 6, 10, 20`
- 中序：`2, 4, 6, 8, 10, 20`
- 後序：`2, 6, 4, 20, 10, 8`


### 3.1 遞迴遍歷示範


````python
def preorder(node):
    if not node:
        return
    print(node.value, end=' ')
    preorder(node.left)
    preorder(node.right)

print('Preorder:')
preorder(root)
print()
````

````python
def inorder(node):
    if not node:
        return
    inorder(node.left)
    print(node.value, end=' ')
    inorder(node.right)

print('Inorder:')
inorder(root)
print()
````

````python
def postorder(node):
    if not node:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.value, end=' ')

print('Postorder:')
postorder(root)
print()
````

### 3.2 層序遍歷 (Level-order)
層序遍歷可以視為對樹使用 BFS。


````python
from collections import deque

def level_order(node):
    if not node:
        return
    queue = deque([node])
    while queue:
        curr = queue.popleft()
        print(curr.value, end=' ')
        if curr.left:
            queue.append(curr.left)
        if curr.right:
            queue.append(curr.right)

print('Level-order:')
level_order(root)
print()
````

## 4. 常見題型與應用
1. **計算樹的高度**：遞迴回傳左右子樹高度 + 1。
2. **計算節點數、葉節點數**：遞迴累加或判斷。
3. **二元搜尋樹 (BST)**：中序遍歷會得到遞增序列。
4. **表達式樹**：節點存放運算子，葉節點存放運算數，可用後序遍歷求值。


### 4.1 練習：計算高度與節點數


````python
def height(node):
    if not node:
        return -1  # 定義空樹高度為 -1，使葉子高度為 0
    return 1 + max(height(node.left), height(node.right))

def count_nodes(node):
    if not node:
        return 0
    return 1 + count_nodes(node.left) + count_nodes(node.right)

print('Height:', height(root))
print('Node count:', count_nodes(root))
````
