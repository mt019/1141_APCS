## 1. Base Case 與遞迴關係
以倒數計時為例：


```python
def countdown(n):
    if n == 0:
        print('發射!')
        return
    print(n)
    countdown(n - 1)

countdown(3)
```

### 思考
- Base case 在哪裡？
- 若移除 base case 會發生什麼事？


## 2. 階乘練習
`n! = n * (n-1)!`。先寫出公式，再實作函式。


```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

## 3. 互遞迴觀察
以下程式與檢測題 Q8 類似，請追蹤呼叫步驟。


```python
def f1(m):
    if m > 3:
        print('f1-stop', m)
        return
    print('f1', m)
    f2(m + 2)
    print('f1-back', m)

def f2(n):
    if n > 3:
        print('f2-stop', n)
        return
    print('f2', n)
    f1(n - 1)
    print('f2-back', n)

f1(1)
```

## 4. GCD 與遞迴
最大公因數（Greatest Common Divisor, 縮寫 GCD）表示兩個數字共同擁有的最大整數因數。
例如 `GCD(18, 24) = 6`，因為 18 與 24 都能被 6 整除，而且沒有更大的共同因數。

### 4.1 輾轉相除法 (Euclidean Algorithm) 的觀念鋪墊
1. 若 `b` 能整除 `a`（即 `a % b == 0`），那 `GCD(a, b)` 就是 `b`。
2. 若 `a` 無法被 `b` 整除，將 `a` 除以 `b`，得到餘數 `r`，則 `GCD(a, b)` 等同於 `GCD(b, r)`。
   - 理由：`a = b * q + r`，任何同時整除 `a` 與 `b` 的數，也一定會整除 `r`。
3. 持續重複第 2 步，餘數會越來越小，最終一定會遇到餘數為 0 的情況，此時當前的除數就是 GCD。

### 4.2 為什麼適合用遞迴？
- 遞迴函式自然對應「持續把問題縮小」的過程：
  1. `GCD(a, b)` 呼叫 `GCD(b, a % b)`。
  2. 問題尺寸在餘數為 0 時終止（Base Case）。
- 每一層呼叫都更靠近答案，不需要額外的迴圈或暫存陣列。

理解這些概念後，再看下方程式碼會更直覺。


```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print(gcd(76, 48))
```
