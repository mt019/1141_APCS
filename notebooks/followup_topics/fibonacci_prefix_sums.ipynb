## 1. 費式數列暖身
費式數列（Fibonacci sequence）是最常見的遞迴數列之一：
- 以 `F(0) = 0`, `F(1) = 1` 作為起始值。
- 對於 `n >= 2`，定義為 `F(n) = F(n-1) + F(n-2)`。

常見前幾項為：`0, 1, 1, 2, 3, 5, 8, 13, ...`，每一項等於前兩項的和。
這個數列會在動態規劃、數學歸納與演算法分析中屢次出現，也能對應到兔子繁殖、斐波那契螺旋等生活例子。

## 2. 迴圈版 Fibonacci
透過暫存前兩項的方式計算第 N 項。


```python
def fib_iter(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    prev, curr = 0, 1
    for _ in range(2, n + 1):
        prev, curr = curr, prev + curr
    return curr

for n in range(6):
    print(n, fib_iter(n))
```

## 3. 線上總和 (Prefix Sum)
建立一個 `prefix` 陣列，`prefix[i]` 代表前 i 個數字的總和。


```python
numbers = [3, 5, 2, 7, 4]
prefix = [0]
for value in numbers:
    prefix.append(prefix[-1] + value)

print('原始資料:', numbers)
print('前綴和:', prefix)

# 範例：計算索引 1~3 的總和 (5 + 2 + 7)
left, right = 1, 3
segment_sum = prefix[right + 1] - prefix[left]
print(f'索引 {left}~{right} 的和 =', segment_sum)
```
