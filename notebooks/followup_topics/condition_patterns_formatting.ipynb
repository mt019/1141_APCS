## 2. 星號圖形輸出
練習使用巢狀迴圈與格式化控制空白與星號：


```python
def print_triangle(rows):
    for i in range(rows):
        spaces = ' ' * (rows - i - 1)
        stars = '*' * (2 * i + 1)
        print(spaces + stars)

print_triangle(4)
```

### 延伸挑戰
- 根據輸入列數與字元，動態生成圖形。
- 嘗試輸出菱形或鑽石形狀。


```python
def interactive_triangle():
    while True:
        try:
            rows = int(input("Enter the number of rows for the triangle (or a negative number to exit): "))
            if rows < 0:
                print("Exiting the program.")
                break
            print_triangle(rows)
        except ValueError:
            print("Please enter a valid integer.")

interactive_triangle()
```
