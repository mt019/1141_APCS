## 什麼是遞迴？

> Recursion is the process in which a function calls itself directly or indirectly.

- 中文常翻為「遞迴」或「遞歸」，概念都是 "用自己定義自己"。
- 遞迴思考方式：把大問題拆成一個容易處理的小部分，以及 "把剩下的交給同樣的函式"。
- 生活類比：俄羅斯娃娃、鏡子對著鏡子、Google 地圖放大到最後都是同一種結構。

## 遞迴函式的兩個核心元素

1. **Base case (停止條件)**：當問題已經小到能直接回答時，要給出直接結果。
2. **Recursive case (拆解/遞迴呼叫)**：把問題拆成較小或較簡單的子問題，再呼叫自己來處理。

只要兩者都正確設計，遞迴就會越走越接近 Base case，最終停止。沒有 Base case 或拆解不正確，就會造成無窮遞迴並導致程式錯誤。

## 與迴圈相比：遞迴的思維方式

| 面向 | 迴圈 Iteration | 遞迴 Recursion |
|------|----------------|----------------|
| 終止條件 | 透過變數/索引逐步改變 | 透過 Base case 明確定義 |
| 解題策略 | 累加或重複執行固定流程 | 將問題拆成同型子問題 |
| 程式狀態 | 依賴變數紀錄狀態 | 呼叫堆疊保存每一層狀態 |
| 優點 | 執行效率通常較高；避免呼叫堆疊 | 可讀性高，貼近數學歸納與 Divide & Conquer 思維 |
| 缺點 | 程式碼可能較繁瑣 | 過深時可能 Stack Overflow，也較難理解效能 |

**關鍵概念：** 遞迴與迴圈都能完成同樣的事情，差別在 *思考方式與程式表達*。

## 第一個例子：倒數計時

從一個簡單的倒數範例理解 Base case 與 Recursive case。

````python
def countdown(n):
    # 遞迴倒數：印出 n, n-1, ..., 1, 然後 Lift off!
    if n <= 0:
        print('Lift off!')  # Base case: 當 n <= 0 時停止
        return

    print(n)  # 工作：印出目前的數字
    countdown(n - 1)  # Recursive case: 把剩下的工作交給較小的問題
````

執行 `countdown(3)` 時，呼叫堆疊會經歷：

```
countdown(3) -> print 3 -> countdown(2)
countdown(2) -> print 2 -> countdown(1)
countdown(1) -> print 1 -> countdown(0)
countdown(0) -> print 'Lift off!' -> return
```

每一次呼叫都會暫時停住，等待下一層結束後再返回。這種先深入、再逆序返回的流程稱為 **呼叫堆疊 (call stack)**。

````python
countdown(5)
````

### 觀察呼叫堆疊的順序

我們可以在進入與離開遞迴時印出提示，輔助追蹤。

````python
def countdown_trace(n):
    print(f'>>> enter countdown({n})')
    if n <= 0:
        print('Lift off!')
    else:
        countdown_trace(n - 1)
        print(f'<<< leave countdown({n})')

countdown_trace(3)
````

## 經典數學例子：階乘 factorial

數學定義：

```
n! = 1                if n = 0 or n = 1
n! = n * (n - 1)!     if n > 1
```

這個定義天生就是遞迴：為了算 `n!`，我們只需要知道 `(n-1)!`。

在程式中要注意兩件事：

1. Base case：`0!` 與 `1!` 都是 `1`。
2. 逐步縮小問題：遞迴呼叫 `factorial(n - 1)`。

````python
def factorial(n):
    if n < 0:
        raise ValueError('n 必須是非負整數')
    if n in (0, 1):
        return 1  # Base case
    return n * factorial(n - 1)  # Recursive case

print(factorial(5))
````

### 手動畫出 `factorial(5)` 呼叫順序

```
factorial(5)
= 5 * factorial(4)
= 5 * (4 * factorial(3))
= 5 * 4 * (3 * factorial(2))
= 5 * 4 * 3 * (2 * factorial(1))
= 5 * 4 * 3 * 2 * 1
= 120
```

從上面的文字堆疊可以看到遞迴在“回來”的時候才會進行乘法。

## 常見錯誤

- **Base case 遺漏或條件不正確**：沒有停止條件會造成無窮遞迴。
- **問題沒有縮小**：如果遞迴呼叫沒有朝 Base case 靠近，就會無止境呼叫自己。
- **忽略函式回傳值**：遞迴常會把子結果結合回傳，要記得 `return`。

## Fibonacci 斐波那契數列

定義如下：

```
F(0) = 0
F(1) = 1
F(n) = F(n-1) + F(n-2)   for n >= 2
```

遞迴實作非常貼近定義，但也因此隱含了大量重複計算。

````python
def fibonacci(n):
    if n < 0:
        raise ValueError('n 必須是非負整數')
    if n in (0, 1):
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
````

### 遞迴樹 (Recursion tree) 小提醒

```
F(5)
├── F(4)
│   ├── F(3)
│   │   ├── F(2)
│   │   │   ├── F(1)
│   │   │   └── F(0)
│   │   └── F(1)
│   └── F(2)
│       ├── F(1)
│       └── F(0)
└── F(3)
    ├── F(2)
    │   ├── F(1)
    │   └── F(0)
    └── F(1)
```

可以看到 `F(2)` 被重複計算了非常多次，這代表時間複雜度呈指數成長。

### 使用記憶化 (Memoization) 提升效率

- 思路：把已經算過的結果存起來，下次直接取用。
- Python 提供 `functools.lru_cache` 可以快速達成。

````python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci_fast(n):
    if n < 0:
        raise ValueError('n 必須是非負整數')
    if n in (0, 1):
        return n
    return fibonacci_fast(n - 1) + fibonacci_fast(n - 2)

print(fibonacci_fast(30))
````

## 遞迴處理資料：列表加總

將問題拆解為：第一個元素 + 剩餘元素的加總。

````python
def recursive_sum(items):
    if not items:  # Base case: 空列表的加總為 0
        return 0
    return items[0] + recursive_sum(items[1:])

print(recursive_sum([3, 5, 7, 9]))
````

### 延伸：展開巢狀列表

透過遞迴，我們可以處理不規則的結構，例如巢狀列表。

````python
def flatten(nested):
    result = []

    def helper(items):
        for item in items:
            if isinstance(item, list):
                helper(item)  # 遞迴展開內部的列表
            else:
                result.append(item)

    helper(nested)
    return result

print(flatten([1, [2, 3], [4, [5, 6]]]))
````

## 遞迴與數學歸納法

數學歸納法（Mathematical Induction）是一種用來證明「對所有自然數都成立」命題的方法，而遞迴的正確性常可以用同樣的思路說明。

- **Base case**：證明最小的輸入（通常是 0 或 1）時命題成立，對應到遞迴中的停止條件。
- **Inductive hypothesis**：假設當輸入較小的值時命題成立，對應到「假設遞迴呼叫可以正確處理較小的子問題」。
- **Inductive step**：在假設成立的前提下，證明當前輸入也成立，對應到遞迴函式如何利用子問題的解構建出答案。

### 範例：證明遞迴求和函式的正確性

假設我們寫了一個函式計算 `1 + 2 + ... + n`。

1. Base case：當 `n = 0` 時回傳 `0`，命題成立。
2. 歸納假設：假設 `sum_to(n-1)` 可以得到 `1 + ... + (n-1) = (n-1)n/2`。
3. 歸納步驟：`sum_to(n)` 回傳 `n + sum_to(n-1)`。根據假設，得到 `n + (n-1)n/2 = n(n+1)/2`，因此命題對 `n` 也成立。

透過歸納法，我們就能有系統地證明遞迴程式的正確性，而不是只靠直覺。

````python
def sum_to(n):
    if n == 0:
        return 0
    return n + sum_to(n - 1)

for n in range(6):
    expected = n * (n + 1) // 2
    result = sum_to(n)
    print(f"n={n:2d} -> sum_to(n)={result}, formula={expected}")
    assert result == expected
````

## 常見遞迴解題模式

遞迴的關鍵在於抓出子問題與原問題的「自相似」結構，常見的套路有：

- **Divide and Conquer（分而治之）**：將問題拆成數個較小、相同型態的子問題，分別解完後整合結果。典型例子有 Merge Sort、Binary Search、快速排序。
- **Backtracking（回溯）**：逐層嘗試所有選項，當某條路不通就回退並改走其他分支，常見於排列組合、迷宮、N 皇后等需要搜遍狀態空間的問題。
- **Tree / Graph Traversal（樹與圖的走訪）**：樹或圖的每個子節點都形成一個更小的子結構，使用遞迴可以簡潔地描述前序、後序、深度優先搜尋等操作。
- **Divide-and-Accumulate（拆解後組合結果）**：處理資料結構時，先對子結構遞迴取得答案，再把當前節點的結果與子結果合併。展開巢狀列表、計算節點總數都屬此類。

掌握這些模式，有助於快速判斷一個問題是否適合用遞迴描述。

### Divide & Conquer 思維詳解

Divide & Conquer（分而治之）通常包含三個步驟：

1. **Divide**：把規模為 `n` 的問題拆成 `a` 個較小的子問題，常見情況是每個子問題大小約為 `n/b`。
2. **Conquer**：遞迴地解決這些子問題，直到命中 base case。
3. **Combine**：將子問題的解整合，還原成原問題的答案。

如果把每一層的工作量寫成遞迴式 `T(n) = a T(n/b) + f(n)`，就可以利用 **Master Theorem** 或遞迴樹（recursion tree）推估時間複雜度。分析 Divide & Conquer 的關鍵在於理解分裂的數量、規模，以及合併時的成本。

#### 範例：Merge Sort 的遞迴流程

Merge Sort 透過以下步驟排序：

1. Divide：將陣列一分為二。
2. Conquer：遞迴排序左右子陣列。
3. Combine：將已排序的左右子陣列用線性時間合併。

遞迴樹示意（`n = 8`）：

```
[1..8]
├── [1..4]
│   ├── [1..2]
│   │   ├── [1]
│   │   └── [2]
│   └── [3..4]
│       ├── [3]
│       └── [4]
└── [5..8]
    ├── [5..6]
    │   ├── [5]
    │   └── [6]
    └── [7..8]
        ├── [7]
        └── [8]
```

每一層合併的成本為 `O(n)`，遞迴深度約為 `log n` 層，因此整體複雜度為 `O(n log n)`。

````python
def merge_sort(items):
    if len(items) <= 1:
        return items

    mid = len(items) // 2
    left = merge_sort(items[:mid])
    right = merge_sort(items[mid:])

    merged = []
    i = j = 0
    # Combine step: 合併兩個已排序的子陣列
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged

example = [38, 27, 43, 3, 9, 82, 10]
print(merge_sort(example))
assert merge_sort(example) == sorted(example)
````

## Divide and Conquer 範例：二分搜尋

二分搜尋每次把列表對半砍，直到找到目標或列表為空。

````python
def binary_search(sorted_list, target, left=0, right=None):
    if right is None:
        right = len(sorted_list) - 1

    if left > right:
        return -1  # Base case: 沒找到

    mid = (left + right) // 2
    guess = sorted_list[mid]

    if guess == target:
        return mid
    if guess > target:
        return binary_search(sorted_list, target, left, mid - 1)
    return binary_search(sorted_list, target, mid + 1, right)

print(binary_search([1, 3, 4, 6, 8, 10, 13], 8))
````

## Backtracking 範例：河內塔 (Tower of Hanoi)

規則：
- 一次只能移動一個圓盤。
- 不能把大圓盤放在小圓盤上面。
- 目標是把全部圓盤從起始柱移到目標柱。

遞迴策略：
1. 先把前 `n-1` 個圓盤移到暫存柱。
2. 把第 `n` 個（最大的）圓盤移到目標柱。
3. 再把暫存柱的 `n-1` 個圓盤移到目標柱。

````python
def hanoi(n, start, target, helper, moves=None):
    if moves is None:
        moves = []

    if n == 1:
        moves.append((start, target))
        return moves

    hanoi(n - 1, start, helper, target, moves)
    moves.append((start, target))
    hanoi(n - 1, helper, target, start, moves)
    return moves

print(hanoi(3, 'A', 'C', 'B'))
````

## 回顧：何時選擇遞迴？

- 問題具有明顯的自相似結構或樹狀結構。
- 需要描述「做一件事，剩下交給自己」的流程。
- 以可讀性或數學證明為優先時。

若追求效能，或知道遞迴深度會很大，則更適合使用迴圈或其他資料結構。
