## 1. 布林語意暖身（Fact vs Statement）

### 活動流程
1. 準備卡片：一半寫事實描述（Fact），另一半寫「可以判真假的句子」（Statement）。
2. 請學生將卡片分類；確認每張 Statement 後面能接 `True`/`False`，Fact 則是用來判斷的依據。
3. 讓學生選出其中兩個 Statement（A、B），建立真值表並把每一列用生活語意描述。
4. 進一步提問：`not A`、`A and B`、`not A or B` 各代表哪些情境？

### Notebook 示範
- 修改下方範例中的 `statements` 與 `facts`，投影輸出供學生參考。


````python
statements = {
    "A": "今天正在下雨",
    "B": "我有帶傘出門",
}

facts = [
    {"A": False, "B": False},
    {"A": True, "B": False},
    {"A": False, "B": True},
    {"A": True, "B": True},
]

for row in facts:
    a_val, b_val = row["A"], row["B"]
    description = []
    description.append(statements["A"] if a_val else f"不是：{statements['A']}")
    description.append(statements["B"] if b_val else f"不是：{statements['B']}")
    print(f"A={a_val:5}  B={b_val:5} ｜ {'，'.join(description)}")
    print(f"  A and B → {a_val and b_val}")
    print(f"  not A or B → {((not a_val) or b_val)}\n")
````

### 教學提示
- 若學生仍把布林變數視為「名詞」，可請他們將 Statement 改寫成「如果…那就…」的完整句子。
- 可加入第三個 Statement（C）讓學生練習括號與優先順序。

### Notebook 活動模板
- 使用下方程式生成空白表格，課堂中可即時修改 `scenarios` 列表並填寫解答。


````python
from itertools import product

scenarios = [
    {
        "A": "今天有數學考試",
        "B": "我帶了計算機",
    },
    {
        "A": "晚餐要吃火鍋",
        "B": "冰箱還有青菜",
    },
    {
        "A": "晚上有排練",
        "B": "功課已經寫完",
    },
]

for idx, scenario in enumerate(scenarios, start=1):
    print(f"=== 情境 {idx} ===")
    for combo in product([False, True], repeat=2):
        a_val, b_val = combo
        result_and = a_val and b_val
        result_or = a_val or b_val
        result_not_a_or_b = (not a_val) or b_val
        print(f"A={a_val:5}  B={b_val:5} ｜ A and B → {result_and} ｜ A or B → {result_or} ｜ not A or B → {result_not_a_or_b}")
    print()
````

## 4. Python 括號符號補充

在課堂中學生對 `[]` 與 `{}` 的用途提出疑問，本單元整理 Python 常見括號的語意：

- `()`：建立 *tuple*（不可變序列）、包覆函式參數或運算優先順序。
- `[]`：建立 *list*（可變序列），也用於依索引或切片存取序列中的元素。
- `{}`：建立 *dict*（鍵值對）或 *set*（不重複集合）；依照內容是否為 `key: value` 決定資料型態。

教學時可強調：括號不僅是外型不同，還代表資料結構的行為差異（是否可變、是否有索引、是否允許重複）。


````python
# 括號示例：逐一展示常見資料結構與操作
items = ['raincoat', 'umbrella', 'boots']      # list 使用 []，可修改、可依索引取值
coords = (25.0, 121.5)                        # tuple 使用 ()，代表固定長度的點
profile = {'name': 'Iris', 'tries': 2}        # dict 使用 {} 搭配 key: value
tags = {'python', 'loop', 'boolean'}          # set 使用 {} 放值，沒有重複與順序

print('list 索引第 1 個元素 =>', items[0])
items.append('gloves')
print('list 新元素 =>', items)

lat, lng = coords
print('tuple 拆解 =>', lat, lng)
# coords[0] = 30  # 解除註解會引發 TypeError，tuple 不可改

print('dict 取值 =>', profile['name'])
profile['tries'] += 1
print('dict 更新後 =>', profile)

tags.add('login')
print('set 加入元素 =>', tags)
````

## 5. 迴圈控制專題

此段將介紹 `break`、`continue`、`pass`、`while/for ... else`，並提供錯誤案例。可以邊跑程式邊讓學生預測輸出。

### 4.1 正向範例


````python
numbers = [5, 3, -1, 2, 0, 4]

print("[break] 遇到第一個負數就停止")
for n in numbers:
    if n < 0:
        print(f"遇到 {n} → break")
        break
    print(f"處理 {n}")
else:
    print("for-else：沒有 break 才會執行")

print("\n[continue] 略過負數但繼續迴圈")
for n in numbers:
    if n < 0:
        print(f"跳過 {n}")
        continue
    print(f"保留 {n}")

print("\n[pass] 先佔位，之後再補邏輯")
for n in numbers:
    if n == 0:
        pass
    else:
        print(f"處理 {n}")
````

### 5.2 常見錯誤示範

- 在 `continue` 之前忘記更新計數器。
- `break` 之後沒有補上其他處理，導致資訊遺漏。
- `pass` 被誤認為會自動跳出或繼續。


````python
print("[錯誤案例] continue 前忘記計數器更新")
count = 0
i = 0
while i < len(numbers):
    if numbers[i] < 0:
        print(f"遇到負數 {numbers[i]}，但忘記 i += 1 → 無限迴圈")
        # continue 之前沒有 i += 1
        break
    count += 1
    i += 1

print("\n[修正版本]")
count = 0
i = 0
while i < len(numbers):
    if numbers[i] < 0:
        i += 1
        continue
    count += 1
    i += 1

print("處理到的正數數量：", count)
````

## 6. 密碼錯誤次數限制：多種版本對照

此段整理四種常見寫法與比較，並搭配模擬輸入函式。除了教師示範之外，建議讓學生兩兩對照差異，填寫比較表。（可參考 `week05_supplement.ipynb` 中的版本。）


````python
def make_mock_input(answers):
    """模擬 input()，按照給定答案依序回傳。"""
    answers = list(answers)
    def mock_input(prompt):
        if not answers:
            raise RuntimeError("沒有更多模擬輸入，請擴充答案列表。")
        value = answers.pop(0)
        print(f"{prompt}{value}")
        return value
    return mock_input
````

````python
def login_version_pre_input(mock_input, password="apcs", max_tries=3):
    """版本 A：先在 while 前輸入一次，容易出現 off-by-one。"""
    tries = 0
    user_input = mock_input("第一次輸入：")
    while user_input != password and tries != max_tries:
        tries += 1
        remaining = max_tries - tries
        print(f"錯誤 {tries} 次，剩餘 {remaining} 次")
        user_input = mock_input("請重新輸入：")
    if tries == max_tries:
        print("登入失敗（多給一次的錯誤案例）")
        return False
    print("登入成功")
    return True

def login_version_double_condition(mock_input, password="apcs", max_tries=3):
    """版本 B：while 同時檢查密碼與次數。"""
    user_input = mock_input("請輸入密碼：")
    tries = 1
    while user_input != password and tries < max_tries:
        remaining = max_tries - tries
        print(f"錯誤，剩餘 {remaining} 次")
        user_input = mock_input("再試一次：")
        tries += 1
    if user_input == password:
        print("登入成功")
        return True
    print("登入失敗（用完所有次數）")
    return False

def login_version_with_break(mock_input, password="apcs", max_tries=3):
    """版本 C：while + break。"""
    tries = 0
    while tries < max_tries:
        user_input = mock_input(f"第 {tries + 1} 次輸入：")
        if user_input == password:
            print("登入成功")
            return True
        tries += 1
        remaining = max_tries - tries
        print(f"錯誤，剩餘 {remaining} 次")
    print("登入失敗")
    return False

def login_version_for_loop(mock_input, password="apcs", max_tries=3):
    """版本 D：for 迴圈。"""
    for attempt in range(1, max_tries + 1):
        user_input = mock_input(f"第 {attempt} 次輸入：")
        if user_input == password:
            print("登入成功")
            return True
        remaining = max_tries - attempt
        if remaining:
            print(f"錯誤，還有 {remaining} 次")
        else:
            print("錯誤，沒有剩餘次數")
    print("登入失敗（for-else 可補充）")
    return False
````

````python
login_version_double_condition(
    make_mock_input(["0000", "1111", "2222", "apcs"]),
    password="apcs",
    max_tries=3,
)
````
