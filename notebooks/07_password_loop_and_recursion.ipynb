## 2. `while` 迴圈與錯誤次數限制
以下程式示範如何限制嘗試次數，超過次數會終止流程：

````python
MAX_TRIES = 5
password = "1234"
tries = 0

user_input = input("請輸入密碼：")

while user_input != password and tries < MAX_TRIES - 1:
    tries += 1
    print(f"密碼錯誤！已錯 {tries} 次")
    user_input = input("請再試一次：")

if user_input == password:
    print("密碼正確，歡迎光臨")
else:
    print("輸入錯誤次數過多，請稍後再試")
````

> 練習：調整訊息讓使用者知道剩餘次數，或改為 `while True` + `break` 的寫法。

## 3. 迴圈控制關鍵字：`break`、`continue`、`return`

### 3.1 `break`：提前結束迴圈
找到正確答案後立刻跳出迴圈，`while ... else` 的 `else` 不會執行：

````python
attempts = 0
while attempts < 5:
    user_input = input("請輸入密碼（輸入 1234 通過）：")
    attempts += 1

    if user_input == "1234":
        print("成功登入")
        break

    print("密碼錯誤")
else:
    print("輸入錯誤次數過多")
````

### 3.2 `continue`：略過當次迴圈剩餘內容
以下程式跳過空字串或預設密碼，僅列出需要檢查的候選：

````python
candidates = ['0000', '', '1111', '1234', '4321']
for guess in candidates:
    if not guess:  # 空字串
        continue
    if guess == '1234':  # 已知正確密碼
        continue
    print('嘗試預設密碼：', guess)
````

### 3.3 `return`：離開函式（可在迴圈內觸發）
`return` 會終止整個函式，常用於成功或失敗就立即返回結果：

````python
def verify(password):
    while True:
        user_input = input("請輸入密碼（輸入 quit 可離開）：")
        if user_input == password:
            return True
        if user_input == 'quit':
            return False
        print('再試一次，或輸入 quit 離開')

result = verify('1234')
print('驗證結果：', result)
````

> 練習：比較 `break` 與 `return` 版本的差異，思考要在哪裡放 `return` 才合理。

## 4. 函式呼叫與值傳遞

### 4.1 函式只會看到傳入的值
觀察函式內外的變數值：

````python
def add_two(x):
    print('函式內收到的 x =', x)
    return x + 2

m = 3
new_value = add_two(m + 2)
print('返回值 =', new_value)
print('函式呼叫後的 m =', m)
````

### 4.2 重新指派才會改變原變數
透過回傳值更新變數：

````python
m = add_two(m)
print('重新指派後的 m =', m)
````

## 5. 遞迴追蹤工具
下方提供簡單的追蹤表格模板，可自行複製填寫：

| 呼叫層級 | 函式 | 參數值 | 動作 | 備註 |
|---------|------|--------|------|------|
| 1 | f1 | m=1 | 進入 | 呼叫 f2(3) |
| 2 | f2 | n=3 | 進入 | 呼叫 f1(2) |
| 3 | f1 | m=2 | ... | ... |

```python
def countdown(n):
    if n <= 0:
        print('Lift off!')
        return
    print(n)
    countdown(n - 1)

countdown(3)
```
